````python
# TODO
# 1. ì•ì—ì„œ ë§Œë“¤ì–´ë‚¸ `continuous_cols`ë¥¼ ì´ìš©í•˜ì—¬ ì—°ì†í˜• ë³€ìˆ˜ë§Œ ë‹´ê²¨ì ¸ìˆëŠ” `X_raw`ë¥¼ ë§Œë“¤ì–´ì£¼ì„¸ìš”.
continuous_cols = df.select_dtypes(include=['number']).columns.tolist()
X_raw: np.ndarray = df[continuous_cols].values

# 2. ê° ì—´ì˜ í‰ê· ì„ ê³„ì‚°í•˜ì—¬ muì— í• ë‹¹í•˜ì„¸ìš”.
mu: np.ndarray = X_raw.mean(axis=0)

# 3. ê° ì—´ì˜ í‘œì¤€í¸ì°¨ë¥¼ ê³„ì‚°í•˜ì—¬ sigmaì— í• ë‹¹í•˜ì„¸ìš”.
# í‘œì¤€í™” ê³¼ì •ì—ì„œ ë¶„ëª¨ì— ë“±ì¥í•˜ëŠ” sigmaì— 0ì´ í¬í•¨ë˜ëŠ” ê²½ìš° ë¬¸ì œê°€ ìƒê¹ë‹ˆë‹¤.
# ì¶”ê°€ë¡œ í˜¹ì‹œ ëª¨ë¥¼ ì•ˆì „ ì¥ì¹˜ë¥¼ ì·¨í•´ë³¼ê¹Œìš”? (ë³„ë„ì˜ í…ŒìŠ¤íŠ¸ì½”ë“œëŠ” ì—†ìŠµë‹ˆë‹¤)
sigma: np.ndarray =  X_raw.std(axis=0)
sigma = np.where(sigma == 0, 1, sigma)

# 4. ë¸Œë¡œë“œìºìŠ¤íŒ…ì„ í™œìš©í•´ (X_raw - mu) / sigma í˜•íƒœë¡œ í‘œì¤€í™”ë¥¼ ìˆ˜í–‰í•˜ê³  X_normì— ì €ì¥í•˜ì„¸ìš”.
X_norm: np.ndarray = (X_raw - mu) / sigma


# ì •ë‹µ í™•ì¸ìš© ì½”ë“œì…ë‹ˆë‹¤. ìˆ˜ì •í•˜ì§€ ì•Šê³  ì´ìš©í•´ì£¼ì„¸ìš”
mu_expected    = np.mean(X_raw, axis=0)
sigma_expected = np.std(X_raw, axis=0)
X_norm_expected = (X_raw - mu_expected) / sigma_expected

# ëª¨ì–‘(shape) ê²€ì¦
assert mu.shape     == mu_expected.shape,     f"mu shape mismatch: {mu.shape} vs {mu_expected.shape}"
assert sigma.shape  == sigma_expected.shape,  f"sigma shape mismatch: {sigma.shape} vs {sigma_expected.shape}"
assert X_norm.shape == X_norm_expected.shape, f"X_norm shape mismatch: {X_norm.shape} vs {X_norm_expected.shape}"

# ê°’ ê²€ì¦ (í—ˆìš© ì˜¤ì°¨ ë‚´ì—ì„œ ë¹„êµ)
np.testing.assert_allclose(mu,    mu_expected,    rtol=1e-6, atol=1e-8)
np.testing.assert_allclose(sigma, sigma_expected, rtol=1e-6, atol=1e-8)
np.testing.assert_allclose(X_norm, X_norm_expected, rtol=1e-6, atol=1e-8)

print("âœ… í‰ê· Â·í‘œì¤€í¸ì°¨ ê³„ì‚° ë° í‘œì¤€í™”ê°€ ì˜¬ë°”ë¥´ê²Œ ìˆ˜í–‰ë˜ì—ˆìŠµë‹ˆë‹¤!")
````

  - Î£ì˜ ì—­ìˆ˜ë¡œ Î£âºë¥¼ ë§Œë“¤ê³  ìœ ì‚¬ì—­í–‰ë ¬ ì¡°í•© í›„ $\theta$

## 2.1 ì •ê·œë°©ì •ì‹(Normal Equation)

### 2.1.1 ì„ í˜•íšŒê·€ë€? (Linear Regression)

ê° ìƒ˜í”Œ $(i)$ì— ëŒ€í•´ ì—°ì†í˜• íŠ¹ì„± ë²¡í„° $\mathbf{x}^{(i)} \in \mathbb{R}^n$ì™€ ëª©í‘œê°’ $y^{(i)}$ê°€ ìˆì„ ë•Œ, íšŒê·€ ëª¨ë¸ì€ ë‹¤ìŒê³¼ ê°™ì´ ì‘ì„±í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

$$
\hat y^{(i)}
= \theta_0 + \theta_1\,x^{(i)}_1 + \cdots + \theta_n\,x^{(i)}_n
\quad\Longleftrightarrow\quad
\hat y = \mathbf{X}_b\,\boldsymbol{\theta}
$$

- $\mathbf{X}_b \in \mathbb{R}^{m \times (n+1)}$: ê° í–‰ì´ $[1,\,x_1,\dots,x_n]$ì¸ ì„¤ê³„ í–‰ë ¬  
- $\boldsymbol{\theta} \in \mathbb{R}^{n+1}$: ì ˆí¸ $\theta_0$ì™€ ê°€ì¤‘ì¹˜ $\theta_{1:n}$ì˜ ë²¡í„°  
- $m$: ìƒ˜í”Œ ìˆ˜
- ìš°ë¦¬ì˜ ì˜ˆì‹œì˜ ê²½ìš°, ì—°ì†í˜• ë³€ìˆ˜ë“¤ì— ì–´ë–¤ ì„ í˜•ê³„ìˆ˜ë¥¼ ê³±í•´ì„œ í•©í•˜ë©´ `price`ë¥¼ ì˜ˆì¸¡í•  ìˆ˜ ìˆëŠ”ê°€?ë¥¼ ë³´ëŠ” ê²ƒ.

### 2.1.2 ë¹„ìš© í•¨ìˆ˜

- ì˜¬ë°”ë¥¸ ì„ í˜•ê³„ìˆ˜ë¥¼ ì°¾ê¸° ìœ„í•´ì„œëŠ”, í˜„ì¬ ì„ í˜•ê³„ìˆ˜ê°€ ì–¼ë§ˆë‚˜ ëª¨ë¸ì„ ì˜ ì„¤ëª…í•˜ëŠ”ê°€ ë‚˜íƒ€ë‚´ëŠ” 'ì§€í‘œ'ê°€ í•„ìš”í•©ë‹ˆë‹¤. ë³´í†µì€ ë¹„ìš© í•¨ìˆ˜ (Cost function, $J$)ë¥¼ í†µí•´ ì–¼ë§ˆë‚˜ í˜„ì¬ ì„ í˜•ê³„ìˆ˜ê°€ `price`ë¥¼ ì˜ ì˜ˆì¸¡í•˜ëŠ”ì§€ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
- ì„ í˜•íšŒê·€ì™€ ê°™ì€ íšŒê·€ (regression) ëª¨ë¸ì—ì„œëŠ” í‰ê·  ì œê³± ì˜¤ì°¨(MSE)ë¥¼ ë¹„ìš©í•¨ìˆ˜ë¡œ ë§ì´ ì‚¬ìš©í•©ë‹ˆë‹¤. ìˆ˜ì‹ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤. ì˜ˆì¸¡ê°’ê³¼ ì‹¤ì œê°’ ì‚¬ì´ì˜ ì°¨ë¥¼ ë‚˜íƒ€ë‚´ê¸° ë•Œë¬¸ì—, 0ì— ê°€ê¹Œìš¸ìˆ˜ë¡ í˜„ì¬ ì„ í˜•ê³„ìˆ˜ë“¤ì´ `price`ë¥¼ ì˜ ì˜ˆì¸¡í•œë‹¤ê³  ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

$$
J(\boldsymbol{\theta})
= \frac{1}{2m}\sum_{i=1}^m \bigl(\hat y^{(i)} - y^{(i)}\bigr)^2
= \frac{1}{2m}(\mathbf{X}_b\boldsymbol{\theta} - \mathbf{y})^\mathsf{T}(\mathbf{X}_b\boldsymbol{\theta}-\mathbf{y})
$$

### 2.1.3 ìµœì í™” ì¡°ê±´ â†’ ì •ê·œë°©ì •ì‹

- ê³ ë“±í•™êµ ë•Œ ë°°ìš´ ë¯¸ë¶„ì„ ìƒê°í•´ë´…ì‹œë‹¤. ë¯¸ë¶„ì´ 0ì´ë˜ëŠ” ì§€ì ì€ í•´ë‹¹ í•¨ìˆ˜ì˜ ê°’ì´ ë³€í•˜ì§€ ì•ŠëŠ” ì§€ì ìœ¼ë¡œ, ì¡°ê±´ì— ë”°ë¼ í•¨ìˆ˜ì˜ ìµœëŒ€ê°’ í˜¹ì€ ìµœì†Œê°’ì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤. í˜„ì¬ ìš°ë¦¬ì˜ ë¹„ìš© í•¨ìˆ˜ëŠ” ë³€ìˆ˜ $\theta$ì— ëŒ€í•œ 2ì°¨ í•¨ìˆ˜ì´ê¸° ë•Œë¬¸ì— ë¯¸ë¶„ì„ 0ìœ¼ë¡œ ë§Œë“œëŠ” ì„ í˜•ê³„ìˆ˜ë¥¼ ì°¾ìœ¼ë©´ ìš°ë¦¬ëŠ” ë¹„ìš© í•¨ìˆ˜ë¥¼ ìµœì†Œí™”í•˜ëŠ”(ì¦‰, `price`ë¥¼ ê°€ì¥ ì˜ ì˜ˆì¸¡í•˜ëŠ”) ì„ í˜•ê³„ìˆ˜ë¥¼ ì°¾ì•˜ë‹¤ê³  í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
- ë¯¸ë¶„ì´ 0ì´ ë˜ëŠ” ì§€ì ì€ í•´ì„ì ìœ¼ë¡œ ë‹¤ìŒê³¼ ê°™ì´ í’€ì–´ë´…ì‹œë‹¤.

$$
\frac{\partial J}{\partial \boldsymbol{\theta}}
= \frac{1}{m}\,\mathbf{X}_b^\mathsf{T}(\mathbf{X}_b\boldsymbol{\theta} - \mathbf{y})
= 0
\quad\Longrightarrow\quad
\mathbf{X}_b^\mathsf{T}\mathbf{X}_b \,\boldsymbol{\theta}
= \mathbf{X}_b^\mathsf{T}\mathbf{y}
$$

ë”°ë¼ì„œ í•´ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.   
$$
\boldsymbol{\theta}
= \bigl(\mathbf{X}_b^\mathsf{T}\mathbf{X}_b\bigr)^{-1}\,\mathbf{X}_b^\mathsf{T}\mathbf{y}
$$

ë†€ëê²Œë„ ì„ í˜•ê³„ìˆ˜ëŠ” ì´ë¯¸ ë°ì´í„°ì™€ ì˜ˆì¸¡ê°’ì˜ ì¡°í•©ì„ í†µí•´ ê³„ì‚°ì„ í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

### 2.1.4 í•´ì„ì  í•´ì˜ ì¥Â·ë‹¨ì 

- ì¥ì   
  - ë‹¨ í•œ ë²ˆì˜ í–‰ë ¬ ì—°ì‚°ìœ¼ë¡œ ìµœì  íŒŒë¼ë¯¸í„° ê³„ì‚°í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.   
- ë‹¨ì   
  1. $\mathbf{X}_b^\mathsf{T}\mathbf{X}_b$ê°€ í’€ë­í¬(full rank)ê°€ ì•„ë‹ ë•Œ ì—­í–‰ë ¬ ì¡´ì¬ ë¶ˆê°€  
  2. íŠ¹ì„± ê°œìˆ˜ $n$ì´ í¬ê±°ë‚˜ ìƒ˜í”Œ ìˆ˜ $m$ì— ë¹„í•´ í¬ë©´ $ O(n^3)$ ë³µì¡ë„ë¡œ ëŠë¦¼  
  3. ìˆ˜ì¹˜ì  ë¶ˆì•ˆì •ì„±: ì¡°ê±´ìˆ˜(condition number)ê°€ í¬ë©´ ì‘ì€ ë…¸ì´ì¦ˆì—ë„ ë¯¼ê°  


### ğŸ§‘â€ğŸ’» **ì‹¤ìŠµ**: ì •ê·œë°©ì •ì‹ ìˆ˜í–‰í•´ë³´ê¸°

ë‹¨ê³„ì ìœ¼ë¡œ ì •ê·œë°©ì •ì‹ì„ í†µí•œ ì„ í˜•ê³„ìˆ˜ë¥¼ ì°¾ì•„ë´…ì‹œë‹¤. ì•ì—ì„œ ì •ê·œí™”í•œ `X_norm`ì„ ì‚¬ìš©í•©ì‹œë‹¤.   
ìµœì¢…ì ìœ¼ë¡œ $\theta$ì˜ ìˆ˜ì‹ì˜ ìš”ì†Œë¥¼ í•˜ë‚˜ì”© ë³€ìˆ˜ì— í• ë‹¹í•´ì¤€ í›„ì— í–‰ë ¬ê³±ì„ ì§„í–‰í•˜ë©´ ë©ë‹ˆë‹¤ :)

ì—¬ê¸°ì„œ ê³„ì‚°í•œ `X_b`ëŠ” ë‚¨ì€ ì‹¤ìŠµì—ì„œ ê³„ì† ì‚¬ìš©ë  ì˜ˆì •ì´ë‹ˆ ì—´ì‹¬íˆ ë§Œë“¤ì–´ì£¼ì„¸ìš”!

````python
# TODO: Noneìœ¼ë¡œ ì±„ì›Œì ¸ìˆëŠ” ì½”ë“œë¥¼ ì±„ìš°ë©° ë§ˆë¬´ë¦¬ í•´ë´…ì‹œë‹¤.
# ì´ë²ˆì—ëŠ” ì •ë‹µí™•ì¸ìš© ì½”ë“œë¥¼ ë³„ë„ë¡œ ë‘ì§€ ì•Šì•˜ìœ¼ë‚˜, í›„ì— ì‹œê°í™”ë¥¼ í†µí•´ í™•ì¸í•´ë´…ì‹œë‹¤.

# í¸ì˜ë¥¼ ìœ„í•´ ì•„ê¹Œ ì •ë¦¬í•œ X_normì„ ë‹¤ì‹œ pd.DataFrame í˜•íƒœë¡œ ë§Œë“¤ì–´ì£¼ì—ˆìŠµë‹ˆë‹¤.
import pandas as pd
X_df = pd.DataFrame(X_norm, columns=continuous_cols)

# 1. X_normì—ì„œ featureì¸ Xì™€ ì˜ˆì¸¡í•´ì•¼í•˜ëŠ” ëŒ€ìƒì¸ `price` yë¥¼ ë¶„ë¦¬í•´ë´…ì‹œë‹¤.
X: np.ndarray = X_df.drop(labels="price", axis=1).values
y: np.ndarray = X_df["price"].values

# 2. ì ˆí¸í•­ì„ ì¶”ê°€í•´ì¤ë‹ˆë‹¤.
m, n = X_norm.shape
X_b = np.c_[np.ones((m, 1)), X_norm]        # shape = (m, n+1)

# 3. ìš”ì†Œë³„ë¡œ ê³„ì‚°ì„ ì§„í–‰í•´ë´…ì‹œë‹¤.
# 3-1. ì—­í–‰ë ¬ì˜ ëŒ€ìƒì¸ `X_b^T @ X_b`ë¥¼ êµ¬í•´ XT_Xì— í• ë‹¹í•´ì£¼ì„¸ìš”.
XT_X  = X_b.T @ X_b                         # shape = (n+1, n+1)
# 3-2. ì—­í–‰ë ¬ê³¼ ê³±í•´ì§€ëŠ” X_b^T @ y`ë¥¼ êµ¬í•´ XT_yì— í• ë‹¹í•´ì£¼ì„¸ìš”.
XT_y  = X_b.T @ y                           # shape = (n+1, 1)

# 4. í•´ë¥¼ êµ¬í•´ì£¼ì„¸ìš”.
theta = np.linalg.inv(XT_X) @ XT_y         # shape = (n+1, 1)
````

ìµœì í™”ê°€ ì™„ë£Œëœ ë§¤ê°œë³€ìˆ˜ \thetaë¥¼ í†µí•´ ì´ì œ ì˜ˆì¸¡ì„ ì§„í–‰í•´ë´…ì‹œë‹¤.

````python
# ì‹œê°í™”ì— ì‚¬ìš©í•  ì½”ë“œëŠ” ì œê°€ ë¯¸ë¦¬ ì§œë‘ì—ˆìŠµë‹ˆë‹¤. ìˆëŠ” ê·¸ëŒ€ë¡œ í™œìš©í•´ì£¼ì‹œë©´ ë©ë‹ˆë‹¤ :)
import matplotlib.pyplot as plt
sns.set_theme()

def plot_prediction(y_true: np.ndarray, y_pred: np.ndarray) -> None:
  y_true = y_true.flatten()
  y_pred = y_pred.flatten()
  assert y_true.shape == y_pred.shape, f"Size mismatch between y_true and y_pred"

  fig, ax = plt.subplots(figsize=(6, 4))

  # íšŒê·€ì„ 
  sns.scatterplot(x=y_true, y=y_pred,
                  alpha=0.5, label="Model Prediction", ax=ax)

  # ì´ìƒì ì¸ ì˜ˆì¸¡ì„ 
  sns.lineplot(x=[y.min(), y.max()],
               y=[y.min(), y.max()],
               label="Ideal Regression", linestyle="--", color="red")

  ax.set_xlabel('Actual Price')
  ax.set_ylabel('Predicted Price')
  ax.set_title('Actual vs Predicted Price')
  fig.tight_layout()
````

````python
# TODO
# 5. ë§ˆì§€ë§‰ìœ¼ë¡œ ëª¨ë¸ì„ í‰ê°€í•´ë´…ì‹œë‹¤.
# ì–»ì–´ë‚¸ `theta`ì™€ ë°ì´í„°ì˜ í–‰ë ¬ê³±ì„ í†µí•˜ì—¬ ì˜ˆì¸¡ê°’ì¸ `y_pred`ë¥¼ êµ¬í•´ë³´ê³ 
# ì˜ˆì¸¡ê°’ê³¼ ì‹¤ì œê°’ ì°¨ì´ë¥¼ ë‚˜íƒ€ë‚´ëŠ” MSEë¥¼ ê³„ì‚°í•´ë´…ì‹œë‹¤.
y_pred = X_b @ theta # shape = (m, 1)
mse    = np.mean((y_pred - y)**2)

print("íŒŒë¼ë¯¸í„° Î¸:\n", theta.flatten())
print(f"MSE: {mse:.2f}")
plot_prediction(y_true=y, y_pred=y_pred)
````

ì‹¤ì œë¡œ ì—°êµ¬ë¥¼ ì§„í–‰í•  ë•Œ MSEê°€ 0ì´ ë‚˜ì˜¤ëŠ” ì™„ë²½í•œ ê²½ìš°ëŠ” ì°¾ê¸°ê°€ ì–´ë µìŠµë‹ˆë‹¤. í•´ë‹¹ ë°ì´í„°ëŠ” ì—°ìŠµìš© ë°ì´í„°ì´ë‹¤ë³´ë‹ˆ ìµœì í™”ê°€ ì˜ë˜ëŠ” ê²½ìš°ì…ë‹ˆë‹¤.  

ë³´í†µ MSEë§Œ ë³´ê³  ëª¨ë¸ì´ ì–¼ë§ˆë‚˜ ì˜í•˜ê³  ìˆëŠ”ì§€ ì•Œ ìˆ˜ ìˆì§€ë§Œ,ì–´ë””ê¹Œì§€ë‚˜ í‰ê°€ ì§€í‘œëŠ” í•˜ë‚˜ì˜ ìˆ«ìë¡œ ëª¨ë¸ì˜ ì„±ëŠ¥ì„ ë³´ê³  ì´ì”© ë•Œë¬¸ì— _ë°ì´í„° ë³„ ì˜ˆì¸¡ìƒí™©_ì„ ì•Œê¸°ëŠ” ì–´ë µìŠµë‹ˆë‹¤_. ë”°ë¼ì„œ ëª¨ë¸ í•™ìŠµ ì¤‘ê°„, í›„ì— ì˜ˆì¸¡ê°’ì— ëŒ€í•œ ì‹œê°í™”ë¥¼ ì§„í–‰í•´ë³´ëŠ” ê²ƒì€ ì¤‘ìš”í•©ë‹ˆë‹¤.

â“ ì˜ˆì¸¡ëŒ€ìƒì¸ `price`ë„ ì •ê·œí™”í•´ì•¼í• ê¹Œ?

- ì„ í˜• íšŒê·€ì˜ í•´(Î¸)ëŠ” íƒ€ê¹ƒ $y$ì˜ ìŠ¤ì¼€ì¼ì— ì„ í˜• ë¹„ë¡€í•©ë‹ˆë‹¤.  
- ì •ê·œë°©ì •ì‹ì´ë‚˜ `np.linalg.lstsq`ë¡œ í•´ë¥¼ êµ¬í•  ë•ŒëŠ” íƒ€ê¹ƒì„ ìŠ¤ì¼€ì¼ë§í•˜ì§€ ì•Šì•„ë„ ìˆ˜í•™ì ìœ¼ë¡œ ì „í˜€ ë¬¸ì œê°€ ì—†ìŠµë‹ˆë‹¤.  
- ë‹¤ë§Œ  
  - íƒ€ê¹ƒì´ ë§¤ìš° í° ê°’(ì˜ˆ: ìˆ˜ë°±ë§Œ ë‹¨ìœ„)ì¼ ê²½ìš° ì»´í“¨í„° ìˆ«ì í‘œí˜„ í•œê³„ë¡œ ì¸í•´ ë“œë¬¼ê²Œ ìˆ˜ì¹˜ ì•ˆì •ì„±ì´ ì¡°ê¸ˆ ë–¨ì–´ì§ˆ ìˆ˜ ìˆê³

  ê°€ ë˜ê³ , $\theta_0$ ìì²´ê°€ ì ˆí¸ì´ ì•„ë‹ˆë¼ $\theta_0 \times c$ê°€ ì ˆí¸ì´ì–´ì„œ í•´ì„ì´ ë³µì¡í•´ì§‘ë‹ˆë‹¤.

## 2.2 ìµœì†Œì œê³±ë²•(Least Squares)ë¡œ í•´ ì°¾ê¸°: `np.linalg.lstsq`

### 2.2.1 ì •ê·œë°©ì •ì‹ì˜ ë¶ˆì•ˆì •ì„±
ì„ í˜•íšŒê·€ ë¬¸ì œë¥¼ í’€ ë•Œ í”íˆ ì‚¬ìš©í•˜ëŠ” ë°©ë²•ì€ ì •ê·œë°©ì •ì‹

ì…ë‹ˆë‹¤. í•˜ì§€ë§Œ ì´ ë°©ì‹ì€ ë‹¤ìŒê³¼ ê°™ì€ í•œê³„ê°€ ìˆìŠµë‹ˆë‹¤.

   - í”¼ì²˜ê°€ ë§ê±°ë‚˜ ì¤‘ë³µë ìˆ˜ë¡ $X_b^\mathsf{T} X_b$ê°€ ë¹„ê°€ì—­í–‰ë ¬ì´ ë˜ì–´ ì—­í–‰ë ¬ ê³„ì‚°ì´ ë¶ˆê°€ëŠ¥í•´ì§‘ë‹ˆë‹¤.

---

### 2.2.2 `np.linalg.lstsq`ì˜ í•µì‹¬ ì›ë¦¬

- íŠ¹ì´ê°’ ë¶„í•´(SVD)ë¥¼ ì´ìš©í•˜ì—¬ **ì˜ì‚¬ì—­í–‰ë ¬(pseudoinverse)**ì„ ê³„ì‚°í•©ë‹ˆë‹¤.  
- ì˜ì‚¬ì—­í–‰ë ¬ $X_b^+$ë¥¼ í†µí•´ ìµœì†Œì œê³± í•´ë¥¼ êµ¬í•©ë‹ˆë‹¤. (ê¸°ì¡´ ì •ê·œë°©ì •ì‹ì˜ $\bigl(X_b^\mathsf{T} X_b\bigr)^{-1} X_b^\mathsf{T}$ë¥¼ ëŒ€ì²´)
  
$$ \theta = X_b^+\,y $$

- ìˆ˜ì¹˜ ì•ˆì •ì„±ì´ ë†’ì•„ ê·¹ë‹¨ì ì¸ í”¼ì²˜ ìŠ¤ì¼€ì¼ ì°¨ì´, ìƒê´€ê´€ê³„ ê³¼ë‹¤ ë¬¸ì œë¥¼ ì™„í™”í•©ë‹ˆë‹¤.


### 2.2.3 ìµœì†Œì œê³±ë²•ì˜ ì£¼ìš” ì¥ì 
- $X_b^\mathsf{T} X_b$ì˜ ì—­í–‰ë ¬ì„ ì§ì ‘ êµ¬í•˜ì§€ ì•Šì•„ë„ ë¨  
- íŠ¹ì´(Singular, ëŠ” ê±°ì˜ íŠ¹ì´)í•œ ìƒí™©ì—ì„œë„ í•´ë¥¼ ì‚°ì¶œ.  
- ì”ì°¨(residuals), ë­í¬(rank), íŠ¹ì´ê°’(singular values)ì„ í•¨ê»˜ ë°˜í™˜í•˜ì—¬ ëª¨ë¸ ì§„ë‹¨ì´ ê°€ëŠ¥  

ìœ„ì™€ ê°™ì€ ì´ìœ ë¡œ ì •ê·œë°©ì •ì‹ ê³„ì‚° ëŒ€ì‹  `np.linalg.lstsq`ë¥¼ í™œìš©í•˜ë©´  
ë”ìš± ê²¬ê³ í•˜ê³  ì‹¤ìš©ì ì¸ ì„ í˜•íšŒê·€ í•´ë¥¼ ì–»ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

- ì°¸ê³ : https://numpy.org/doc/stable/reference/generated/numpy.linalg.lstsq.html

````python
# TODO: `lstsq`ë¥¼ ì´ìš©í•˜ì—¬ `theta_lstsq`ë¥¼ êµ¬í•´ì£¼ì„¸ìš”.
theta_lstsq, residuals, rank, singular_vals = np.linalg.lstsq(X_b, y, rcond=None)
````

````python
# TODO
# ë§ˆì§€ë§‰ìœ¼ë¡œ ëª¨ë¸ì„ í‰ê°€í•´ë´…ì‹œë‹¤.
# ì–»ì–´ë‚¸ `theta_lstsq`ì™€ ë°ì´í„°ì˜ í–‰ë ¬ê³±ì„ í†µí•˜ì—¬ ì˜ˆì¸¡ê°’ì¸ `y_pred`ë¥¼ êµ¬í•´ë³´ê³ 
# ì˜ˆì¸¡ê°’ê³¼ ì‹¤ì œê°’ ì°¨ì´ë¥¼ ë‚˜íƒ€ë‚´ëŠ” MSEë¥¼ ê³„ì‚°í•´ë´…ì‹œë‹¤.
y_pred = X_b @ theta_lstsq
mse    = np.mean((y_pred - y)**2)

print("íŒŒë¼ë¯¸í„° Î¸:\n", theta_lstsq.flatten())
print(f"MSE: {mse:.2f}")
plot_prediction(y_true=y, y_pred=y_pred)
````

## 2.3 SVDë¥¼ ì´ìš©í•œ ìµœì†Œì œê³± í•´ êµ¬í•˜ê¸°

SVD(íŠ¹ì´ê°’ ë¶„í•´, Singular Value Decomposition)ëŠ” ì„ì˜ì˜ í–‰ë ¬ $A\in\mathbb{R}^{m\times n}$ë¥¼  

$$
A = U\,\Sigma\,V^\mathsf{T}  
$$

  
ìœ¼ë¡œ ë¶„í•´í•˜ëŠ” ê¸°ë²•ì…ë‹ˆë‹¤. ì—¬ê¸°ì„œ  
- $U\in\mathbb{R}^{m\times r}$ì™€ $V\in\mathbb{R}^{n\times r}$ëŠ” ì§êµ í–‰ë ¬ (orthogonal)
- $\Sigma\in\mathbb{R}^{r\times r}$ëŠ” íŠ¹ì´ê°’ì„ ëŒ€ê°ìœ¼ë¡œ ê°€ì§„ ëŒ€ê° í–‰ë ¬ì´ë©° $r=\mathrm{rank}(A)$ì…ë‹ˆë‹¤.  

SVDë¥¼ ì´ìš©í•˜ë©´ ìµœì†Œì œê³± í•´ë¥¼ ë‹«íŒ í˜•ì‹ìœ¼ë¡œ êµ¬í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. pseudoinverse $\Sigma^+$ë¥¼ ì •ì˜í•˜ë©´  

$$
\Sigma^+_{ii} = \frac{1}{\Sigma_{ii}}\quad(\Sigma_{ii}>0)
$$


ì´ê³ , ì´ë•Œ  


$$ x^* = A^+ b = V\,\Sigma^+\,U^\mathsf{T} b
$$


ë¡œ ìµœì†Œì œê³± í•´ $\min_x\|Ax - b\|_2^2$ë¥¼ ì–»ê²Œ ë©ë‹ˆë‹¤.  

`np.linalg.lstsq`ì™€ ë‹¬ë¦¬ ì§ì ‘ SVD (`np.linalg.svd`)ë¥¼ ì‚¬ìš©í•˜ë©´  
- íŠ¹ì´ê°’ ì ˆë‹¨(threshold)ì„ ì§ì ‘ ì œì–´í•  ìˆ˜ ìˆê³   
- $U, \Sigma, V^\mathsf{T}$ë¥¼ í•™ìŠµì´ë‚˜ ì‹œê°í™”ìš©ìœ¼ë¡œ ë°”ë¡œ í™œìš©í•  ìˆ˜ ìˆìœ¼ë©°  
- ì •ê·œí™”(regularization)ë‚˜ ì €ì°¨ì› ê·¼ì‚¬(rank-truncated) í•´ë²• ì„¤ê³„ê°€ ìˆ˜ì›”í•´ì§‘ë‹ˆë‹¤.
- ì°¸ê³ : https://numpy.org/doc/stable/reference/generated/numpy.linalg.svd.html  

````python
# TODO:
# SVDë¥¼ ìˆ˜í–‰í•˜ì—¬ í•´ë¥¼ êµ¬í•´ë´…ì‹œë‹¤.

# 1. SVDë¥¼ ìˆ˜í–‰í•˜ì—¬ U, S, V^Të¥¼ êµ¬í•´ë´…ì‹œë‹¤.
U, S, Vt = np.linalg.svd(X_b, full_matrices=False)

# 2. Pseudo-inverse S_plusë¥¼ êµ¬í•´ë´…ì‹œë‹¤.
S_plus = np.diag(1.0 / S)

# 3. ì•ì„œ ë°°ìš´ ìˆ˜ì‹ì„ ì°¸ê³ í•˜ì—¬ `theta_svd`ë¥¼ êµ¬í•´ë´…ì‹œë‹¤
theta_svd = Vt.T @ S_plus @ U.T @ y
````

````python
# TODO
# ë§ˆì§€ë§‰ìœ¼ë¡œ ëª¨ë¸ì„ í‰ê°€í•´ë´…ì‹œë‹¤.
# ì–»ì–´ë‚¸ `theta_svd`ì™€ ë°ì´í„°ì˜ í–‰ë ¬ê³±ì„ í†µí•˜ì—¬ ì˜ˆì¸¡ê°’ì¸ `y_pred`ë¥¼ êµ¬í•´ë³´ê³ 
# ì˜ˆì¸¡ê°’ê³¼ ì‹¤ì œê°’ ì°¨ì´ë¥¼ ë‚˜íƒ€ë‚´ëŠ” MSEë¥¼ ê³„ì‚°í•´ë´…ì‹œë‹¤.
y_pred = X_b @ theta_svd
mse    = np.mean((y_pred - y)**2)

print("íŒŒë¼ë¯¸í„° Î¸:\n", theta_svd.flatten())
print(f"MSE: {mse:.2f}")
plot_prediction(y_true=y, y_pred=y_pred)
````

# 3. ê²½ì‚¬ í•˜ê°•ë²•ê³¼ ì†ì‹¤ ê³„ì‚°

### 3.1 ì™œ ê²½ì‚¬ í•˜ê°•ë²•ì„ ì‚¬ìš©í•´ì•¼ í• ê¹Œ?

ì„ í˜•ëŒ€ìˆ˜ ê¸°ë°˜ì˜ ë‹«íŒ í˜•ì‹ í•´ë²•(ì •ê·œë°©ì •ì‹, SVD ë“±)ì€  
- $X_b^\mathsf{T}X_b$ì˜ ì—­í–‰ë ¬ ê³„ì‚° ë˜ëŠ” ì „ì²´ í–‰ë ¬ ë¶„í•´ê°€ í•„ìš”í•©ë‹ˆë‹¤.
- íŠ¹ì„± ìˆ˜($n$)ë‚˜ ë°ì´í„° ìˆ˜ ($m$)ê°€ ì»¤ì§€ë©´ ë©”ëª¨ë¦¬ì™€ ì—°ì‚°ëŸ‰ì´ **ê¸‰ê²©íˆ** ì¦ê°€í•©ë‹ˆë‹¤.  

ê²½ì‚¬ í•˜ê°•ë²•ì€  
- ë°°ì¹˜(batch), ë¯¸ë‹ˆë°°ì¹˜(mini-batch), ì˜¨ë¼ì¸(online) ë°©ì‹ìœ¼ë¡œ ë°ì´í„°ë¥¼ ë‚˜ëˆ„ì–´ ì²˜ë¦¬í•´ **ë©”ëª¨ë¦¬ ì œì•½ì„ ì™„í™”**í•˜ê³  ëŒ€ê·œëª¨ ë°ì´í„°ì— ìœ ì—°í•˜ê²Œ ëŒ€ì‘í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.  
- ì •ê·œí™” í•­ì´ë‚˜ ë¹„ì„ í˜• ì†ì‹¤ í•¨ìˆ˜ì—ë„ ì‰½ê²Œ í™•ì¥ ê°€ëŠ¥í•´ ë”¥ëŸ¬ë‹ ë“± ë„“ì€ ì˜ì—­ì—ì„œ í•µì‹¬ ê¸°ë²•ìœ¼ë¡œ ì“°ì…ë‹ˆë‹¤.  


### 3.2 ì†ì‹¤ í•¨ìˆ˜(MSE) ê³„ì‚°

ëª¨ë¸ ì˜ˆì¸¡ $\hat y = X_b\,\theta$ì™€ ì‹¤ì œ $y$ ì‚¬ì´ì˜ í‰ê·  ì œê³± ì˜¤ì°¨ëŠ” ì•ì„  2.1 ì •ê·œë°©ì •ì‹ì—ì„œ ë‹¤ë£¬ ì†ì‹¤í•¨ìˆ˜ì™€ ë™ì¼í•©ë‹ˆë‹¤.
$$
\mathrm{J(\theta)} = \frac{1}{m}\sum_{i=1}^{m}\bigl(\hat y_i - y_i\bigr)^2
$$

í•™ìŠµ ë°˜ë³µë§ˆë‹¤ MSEë¥¼ ê³„ì‚°í•´ `loss_history`ì— ì €ì¥í•˜ë©´  
- ìˆ˜ë ´ ê³¼ì •ì„ ì‹œê°í™”  
- í•™ìŠµë¥  ìŠ¤ì¼€ì¤„ë§ ë˜ëŠ” ì¡°ê¸° ì¢…ë£Œ(early stopping) ê¸°ì¤€  
ë¡œ í™œìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.  


### 3.3 ê·¸ë˜ë””ì–¸íŠ¸ ê³„ì‚° ë° ì—…ë°ì´íŠ¸
ì„ í˜• íšŒê·€ MSEì˜ íŒŒë¼ë¯¸í„° $\theta$ì— ëŒ€í•œ ê·¸ë˜ë””ì–¸íŠ¸ëŠ”  

$$
\nabla_\theta = \frac{2}{m}\,X_b^\mathsf{T}\bigl(X_b\,\theta - y\bigr)
$$


ì…ë‹ˆë‹¤.  

í•™ìŠµë¥  $\alpha$ë¥¼ ì‚¬ìš©í•œ ì—…ë°ì´íŠ¸ ì‹ì€  


$$
\theta \leftarrow \theta - \alpha\,\nabla_\theta
$$


ë¡œ, ê¸°ìš¸ê¸°ê°€ ê°€ë¦¬í‚¤ëŠ” ì†ì‹¤ ìƒìŠ¹ ë°©í–¥ì˜ ë°˜ëŒ€ë¡œ ì´ë™í•˜ë©´ì„œ ì†ì‹¤ì„ ì ì§„ì ìœ¼ë¡œ ì¤„ì—¬ ê°‘ë‹ˆë‹¤.  


````python
# TODO: Noneìœ¼ë¡œ ì…ë ¥ëœ ë¶€ë¶„ì„ ìš”êµ¬ì‚¬í•­ì— ë§ì¶° ìˆ˜ì •í•˜ë©° ë¬¸ì œë¥¼ í’€ì–´ì£¼ì„¸ìš”
from tqdm import tqdm

# 1. ì´ˆê¸°íŒŒë¼ë¯¸í„° ì„¤ì •
theta        = np.zeros(n+1)
alpha        = 0.01
iterations   = 1000
loss_history = []

# TODO:
# 2. ê²½ì‚¬í•˜ê°•ë²• ë£¨í”„ë¥¼ ë§Œë“¤ì–´ë´…ì‹œë‹¤.
# ì •í•´ì§„ iterations ìˆ˜ë§Œí¼
for i in tqdm(range(iterations)):
    # í˜„ì¬ ë§¤ê°œë³€ìˆ˜ `theta`ì— ì˜í•œ ì˜ˆì¸¡ê°’ `y_pred`ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.
    y_pred     = (X_b @ theta).flatten()

    # ì˜ˆì¸¡í•´ì•¼í•˜ëŠ” ì‹¤ì œê°’ `y`ì™€ì˜ ì°¨ `error`ë¥¼ ê³„ì‚°í•˜ê³ ,
    # `error`ë¥¼ ê¸°ë°˜ìœ¼ë¡œ `mse`ë¥¼ ê³„ì‚°í•´ì£¼ì„¸ìš”.
    error      = y_pred - y
    mse        = np.mean(error**2)
    loss_history.append(mse)

    # ìœ„ì—ì„œ ê³µë¶€í•œ ì‹ì„ í† ëŒ€ë¡œ `gradient`ë¥¼ ê³„ì‚°í•´ì£¼ì„¸ìš”.
    gradient   = (2/m) * X_b.T @ error

    # ì–»ì–´ë‚¸ `gradient`ë¥¼ ê¸°ë°˜ìœ¼ë¡œ `theta`ë¥¼ ì—…ë°ì´íŠ¸í•´ì£¼ì„¸ìš”.
    theta     -= alpha * gradient

# 3. í•™ìŠµ ê²°ê³¼ë¥¼ ì‹œê°í™”í•´ë´…ì‹œë‹¤.
print("ìµœì¢… Î¸:", theta.flatten())
print("ìµœì¢… MSE:", loss_history[-1])

plt.plot(loss_history)
plt.xlabel("Iteration", size="large")
plt.ylabel("Mean Squared Error", size="large")
plt.title("Loss Curve", size="large")
plt.show()
````

ì—¬ê¸°ì„œ ê³ ë¯¼í• ë§Œí•œ ì ì´ ë§ìŠµë‹ˆë‹¤.   
â“ **ì´ˆê¸°ì„¤ì •ê°’, ì €ë ‡ê²Œí•˜ëŠ”ê²Œ ë§ë‚˜ìš”...?**
  - Iterations, í•™ìŠµë¥  (alpha), ì´ˆê¸°ë§¤ê°œë³€ìˆ˜ê°’ ë“± ëª¨ë‘ ì €ë ‡ê²Œ ì´ˆê¸°í™”í•˜ëŠ”ê²Œ ì •ë‹µì¼ê¹Œìš”?
  - ì‚¬ì‹¤ ì •í•´ì§„ ê°’ì€ ì—†ìŠµë‹ˆë‹¤. ì–´ëŠ ì •ë„ ê²½í—˜ìœ¼ë¡œë¶€í„° ì´ˆê¸°ì„¤ì •ë˜ëŠ” ê²ƒë“¤ì´ ë§ìŠµë‹ˆë‹¤. íŠ¹íˆ í•™ìŠµë¥ ì˜ ê²½ìš° í•™ìŠµì—ì„œ ì‚¬ì‹¤ìƒ ê°€ì¥ í° ì˜í–¥ì„ ì¤€ë‹¤ê³  í•´ë„ ê³¼ì–¸ì´ ì•„ë‹ ì •ë„ë¡œ (ê°’ì— ë”°ë¼ ìˆ˜ë ´í•˜ì§€ ì•ŠëŠ” ê²½ìš°ê°€ ë§ì•„ì„œ) ì¤‘ìš”í•©ë‹ˆë‹¤. ê·¸ë ‡ê¸° ë•Œë¬¸ì— ì—¬ëŸ¬ë²ˆ ë§ì€ ì‹¤í—˜ì„ í†µí•´ ì¢‹ì€ ê°’ì„ ì°¾ì•„ì£¼ëŠ” ê²ƒì´ ì¢‹ìŠµë‹ˆë‹¤.
  - Iterationì´ë‚˜ í•™ìŠµë¥ ê³¼ ê°™ì´ _í•™ìŠµë˜ì§€ ì•ŠëŠ”_ ê°’ë“¤ì„ ìš°ë¦¬ëŠ” **ì´ˆë§¤ê°œë³€ìˆ˜ (Hyperparameters)**ë¼ê³  ë¶€ë¦…ë‹ˆë‹¤.

â“ **iterationsì— ë„ë‹¬í•˜ê¸° ì „ì— MSEê°€ ë„ë‹¬í•  ìˆ˜ë„ ìˆì§€ ì•Šë‚˜...?**
  - ë§ìŠµë‹ˆë‹¤. ì¤‘ê°„ì— íŠ¸ë¦¬ê±°ë¥¼ ê±¸ì–´ ëª©í‘œ ì†ì‹¤í•¨ìˆ˜ ê°’ì— ë„ë‹¬í–ˆì„ ë•Œ ë£¨í”„ë¥¼ ê¹¨ê³  í•™ìŠµì„ ë©ˆì¶œ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

### ğŸ§‘â€ğŸ’» **ì‹¤ìŠµ**: ì•Œê³ ë¦¬ì¦˜ ë³´ì™„í•˜ê¸°

ìœ„ì—ì„œ ë‚˜ì˜¨ ì§ˆë¬¸ë“¤ì˜ í•´ë‹µì„ ë°˜ì˜í•´ ë³´ì™„í•´ë´…ì‹œë‹¤. ëª¨ì•„ì„œ í•¨ìˆ˜ë¡œ ë§Œë“¤ì–´ì„œ ì§„í–‰í•´ë´…ì‹œë‹¤.

1. **Early Stopping**: MSEê°€ ì´ë¯¸ 0, í˜¹ì€ 0ì— ê°€ê¹Œìš´ ê°’ì— ìˆ˜ë ´í–ˆë‹¤ë©´ êµ³ì´ í•™ìŠµì„ ë” ì§„í–‰í•  í•„ìš”ê°€ ì—†ìŠµë‹ˆë‹¤. ë”°ë¼ì„œ ì¼ì • MSE ì´í•˜ë¡œ ë–¨ì–´ì¡Œì„ ê²½ìš° ë£¨í”„ë¥¼ ê¹¨ê³  í•™ìŠµì„ ë§ˆë¬´ë¦¬í•˜ëŠ” ì½”ë“œë¥¼ ì¶”ê°€í•´ë´…ì‹œë‹¤.
2. **Mini-batch training**: ë°ì´í„°ê°€ ë„ˆë¬´ ë§ì„ ê²½ìš° ì†ì‹¤ ê³„ì‚°ê³¼ ë¯¸ë¶„ ê³„ì‚°ì— ìˆì–´ ì—°ì‚°ëŸ‰ì´ ë§ì•„ì ¸ ê³„ì‚°ì´ ì–´ë ¤ìš´ ê²½ìš°ê°€ ìˆìŠµë‹ˆë‹¤. ì´ë¥¼ ë³´ì™„í•œ ê²ƒì´ Mini-batch Gradient Descentë¡œ ì „ì²´ ë°ì´í„°ì— ëŒ€í•œ ë¯¸ë¶„ì´ ì•„ë‹Œ ì¼ë¶€ ë°ì´í„°ë§Œì„ ì´ìš©í•œ ë¯¸ë¶„ì„ í†µí•´ ë§¤ê°œë³€ìˆ˜ë¥¼ ì—…ë°ì´íŠ¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì´ _ì¼ë¶€ ë°ì´í„°_ë¥¼ `batch`ë¼ê³  í‘œí˜„í•©ë‹ˆë‹¤. `batch_size`ë¥¼ ì§€ì •í•˜ì—¬ ë‚˜ëˆ ì„œ ì—…ë°ì´íŠ¸ë¥¼ ì§„í–‰í•´ë´…ì‹œë‹¤.
  - ì°¸ê³ : https://medium.com/analytics-vidhya/gradient-descent-vs-stochastic-gd-vs-mini-batch-sgd-fbd3a2cb4ba4
3. **Gradient Accumulation**ì€ ë©”ëª¨ë¦¬ë‚˜ í•˜ë“œì›¨ì–´ ì œì•½ìœ¼ë¡œ í•œ ë²ˆì— í° ë°°ì¹˜ í¬ê¸°ë¥¼ ì‚¬ìš©í•  ìˆ˜ ì—†ì„ ë•Œ, ì‘ì€ ë¯¸ë‹ˆë°°ì¹˜ë“¤ì˜ ê¸°ìš¸ê¸°(gradient)ë¥¼ ëª¨ì•„ì„œ(accumulate) í•œ ë²ˆì— íŒŒë¼ë¯¸í„°ë¥¼ ê°±ì‹ í•˜ëŠ” ê¸°ë²•ì…ë‹ˆë‹¤.  
  - ë³´í†µ `batch_size`ë¥¼ ì¤„ì´ë©´ ë©”ëª¨ë¦¬ëŠ” ì ˆì•½ë˜ì§€ë§Œ, ë„ˆë¬´ ìì£¼ ì—…ë°ì´íŠ¸ê°€ ì¼ì–´ë‚˜ ë…¸ì´ì¦ˆê°€ ì»¤ì§€ê³  í†µì‹  ì˜¤ë²„í—¤ë“œê°€ ì¦ê°€í•©ë‹ˆë‹¤.  
  - `accumulate_steps`ë§Œí¼ì˜ ë¯¸ë‹ˆë°°ì¹˜ ê¸°ìš¸ê¸°ë¥¼ ë”í•´ ë‘ì—ˆë‹¤ê°€, ê·¸ ëˆ„ì ëœ ê¸°ìš¸ê¸°ë¥¼ í•œ ë²ˆì— ë°˜ì˜í•¨ìœ¼ë¡œì¨  
    - íš¨ê³¼ì ìœ¼ë¡  í° `effective_batch_size = batch_size Ã— accumulate_steps`ë¥¼ ì‚¬ìš©  
    - ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰ì€ ë‚®ì¶”ë©´ì„œë„ ì—…ë°ì´íŠ¸ ë¹ˆë„ì™€ ì•ˆì •ì„± ì‚¬ì´ ê· í˜•ì„ ë§ì¶œ ìˆ˜ ìˆìŠµë‹ˆë‹¤.  
  - ë¶„ì‚° í•™ìŠµ í™˜ê²½ì—ì„œëŠ” ê° ì›Œì»¤(worker)ë§ˆë‹¤ ë¯¸ë‹ˆë°°ì¹˜ ê¸°ìš¸ê¸°ë¥¼ ëª¨ì€ ë’¤, ì¤‘ì•™ ì„œë²„ì— í•œ ë²ˆ ì „ì†¡í•´ í†µì‹ ëŸ‰ì„ í¬ê²Œ ì¤„ì´ëŠ” ìš©ë„ë¡œë„ ì“°ì…ë‹ˆë‹¤.



````python
# ì¤‘ê°„ì¤‘ê°„ TODO ë¶€ë¶„ì„ ì±„ìš°ë©° ì½”ë“œë¥¼ ì™„ì„±ì‹œì¼œì£¼ì„¸ìš” :)
def train_linear_regression_sequential(
    X_b, y,
    alpha=0.01,
    epochs=100,
    tol=None,
    batch_size=None,
    accumulate_steps=1,
):
    """
    Shuffling ì—†ì´ ìˆœì°¨ì ìœ¼ë¡œ ë¯¸ë‹ˆë°°ì¹˜ë¥¼ êµ¬ì„±í•˜ëŠ” ì„ í˜• íšŒê·€ í•™ìŠµ í•¨ìˆ˜
    X: (m, n) ì…ë ¥ í”¼ì²˜
    y: (m,) íƒ€ê¹ƒ
    alpha: í•™ìŠµë¥ 
    epochs: ì „ì²´ ë°˜ë³µ íšŸìˆ˜
    tol: MSE ê¸°ì¤€ ì¡°ê¸° ì¢…ë£Œ ì„ê³„ê°’
    batch_size: ë¯¸ë‹ˆë°°ì¹˜ í¬ê¸° (Noneì´ë©´ ì „ì²´ ë°°ì¹˜)
    accumulate_steps: Gradient Accumulationí•  ë°°ì¹˜ ìˆ˜
    """
    m, n = X_b.shape
    theta = np.zeros(n) # íŒŒë¼ë¯¸í„° ì´ˆê¸°í™”
    history = []

    # ì´ë ‡ê²Œ ì„¤ì •í•˜ë©´ batch_sizeë¥¼ ì£¼ì§€ ì•Šì•˜ì„ ë•Œ ì „ì²´ ë°ì´í„° ë¯¸ë¶„ì„ ì‚¬ìš©í•˜ê²Œ ë©ë‹ˆë‹¤
    bs = batch_size or m

    for epoch in range(1, epochs + 1):
        grad_accum = np.zeros_like(theta)
        accum_count = 0

        # ìˆœì°¨ì  ë¯¸ë‹ˆë°°ì¹˜ í•™ìŠµ
        for start in range(0, m, bs):
            # TODO: Mini-batch êµ¬ì„±í•˜ê¸°
            # í˜„ì¬ batch ë‚´ì˜ index ì‹œì‘ ìœ„ì¹˜ë¥¼ ë‚˜íƒ€ë‚´ëŠ” `start` ë³€ìˆ˜ë¥¼ ê¸°ë°˜ìœ¼ë¡œ
            # batchë¥¼ êµ¬ì„±í•´ì¤ë‹ˆë‹¤.
            end = min(start + bs, m)
            X_batch = X_b[start:end]
            y_batch = y[start:end]

            # TODO: Calculating Gradients on mini-batch
            # í˜„ì¬ batchì˜ predictionì„ ê³„ì‚°í•˜ê¸°
            # predictionê³¼ ground truth ì‚¬ì´ ì˜¤ì°¨ ê³„ì‚°í•˜ê¸°
            # ì˜¤ì°¨ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ë¯¸ë¶„ ê³„ì‚°í•˜ê¸°
            y_pred = X_batch.dot(theta)
            error = y_pred - y_batch
            grad  = (2 / X_batch.shape[0]) * X_batch.T.dot(error)

            grad_accum += grad
            accum_count += 1
            # ëˆ„ì ëœ gradientë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì—…ë°ì´íŠ¸
            if accum_count == accumulate_steps or end == m:
                theta -= alpha * (grad_accum / accum_count)
                grad_accum[:] = 0
                accum_count = 0

        # TODO: ì „ì²´ MSE ê³„ì‚° ë° ê¸°ë¡
        # batch ë³„ë¡œ MSEë¥¼ ê³„ì‚°í•  ìˆ˜ ìˆì§€ë§Œ, ì´ëŠ” ì „ì²´ ë°ì´í„°ì— ëŒ€í•œ ê²ƒì´ ì•„ë‹ˆê¸° ë•Œë¬¸ì—
        # ëˆ„ì ëœ ì „ì²´ ë°ì´í„°ì˜ MSEë¥¼ ê³„ì‚°í•´ì¤ë‹ˆë‹¤.
        # ì´ëŠ” batch loopë¥¼ ëŒ ë•Œ MSE ê²°ê³¼ë¥¼ ì €ì¥í•˜ì—¬ ë£¨í”„ í›„ í‰ê· ì„ ë‚´ê±°ë‚˜
        # ì „ì²´ ë°ì´í„°ì— ëŒ€í•´ ë‹¤ì‹œ ì¶”ë¡ ì„ í•˜ëŠ” ë°©ì‹ìœ¼ë¡œ ì§„í–‰í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
        # `print`ë¬¸ì„ í†µí•˜ì—¬ ì¤‘ê°„ì¤‘ê°„ í•™ìŠµê²°ê³¼ë¥¼ ê³„ì† í™•ì¸í•˜ëŠ” ê²ƒì´ ì¤‘ìš”í•©ë‹ˆë‹¤.
        mse = np.mean((X_b.dot(theta) - y) ** 2)
        history.append(mse)
        print(f"Epoch {epoch:3d}/{epochs}    MSE: {mse:.6f}")

        # TODO: Early stopping
        # í˜„ì¬ MSEê³¼ tolì„ ë¹„êµí•˜ì—¬ í•™ìŠµì„ ì§„í–‰í•˜ê±°ë‚˜ Loopë¥¼ ê¹¹ë‹ˆë‹¤
        if tol is not None and mse < tol:
            print(f"-> Early stopping at epoch {epoch}, MSE={mse:.6f}")
            break

    return theta, history
````

````python
theta_est, loss_hist = train_linear_regression_sequential(
    X_b, y,
    alpha=0.01,
    epochs=100,
    tol=1e-4,
    batch_size=32,
    accumulate_steps=2
)
print("ì¶”ì •ëœ Î¸:", theta_est)
````
