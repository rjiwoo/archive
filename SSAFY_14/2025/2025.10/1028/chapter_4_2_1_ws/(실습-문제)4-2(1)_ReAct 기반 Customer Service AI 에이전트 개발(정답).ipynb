## Agent의 도구 가방 채우기 (Tool-learning & Trustworthiness)

이번 과제에서는 우리 에이전트가 고객의 복잡한 민원을 해결하기 위해 사용할 새로운 도구들을 만들고, 이 도구들이 믿을 수 있게 작동하도록 '신뢰성(Trustworthiness)'이라는 개념을 적용해 볼 것입니다.

### 학습 키워드 설명

**Tool-learning (도구 학습)**

Tool-learning은 에이전트가 외부 도구나 함수를 사용하는 방법을 배우는 능력을 말합니다. 마치 사람이 망치, 드라이버 같은 도구를 사용해서 어떤 작업을 하듯이, AI 에이전트도 특정 작업을 수행하기 위해 프로그래밍된 함수나 외부 시스템과 연동되는 도구를 사용할 수 있습니다. 에이전트는 사용자의 요청을 이해하고, 그 요청을 처리하기 위해 어떤 도구가 필요한지 스스로 판단하여 선택하고 사용하는 방법을 학습합니다. 이렇게 하면 에이전트가 자신의 학습 데이터에만 의존하지 않고, 실시간 정보에 접근하거나 특정 기능을 실행하는 등 더 다양하고 복잡한 작업을 수행할 수 있게 됩니다.

**RAG (Retrieval-Augmented Generation, 검색 증강 생성)**

RAG는 '검색(Retrieval)'과 '생성(Generation)'을 결합한 기술입니다. 에이전트가 질문을 받으면, 먼저 가지고 있는 문서나 데이터베이스에서 질문과 관련된 정보를 '검색(Retrieval)'합니다. 그리고 검색된 정보를 참고하여 가장 적절하고 정확한 답변을 '생성(Generation)'해 냅니다. 이렇게 하면 에이전트가 최신 정보나 특정 도메인(영역)의 전문 지식 없이도 정확한 답변을 할 수 있게 됩니다. 우리 프로젝트에서는 배송 정책이나 보상 정책 같은 회사 내부 문서를 RAG를 통해 에이전트에게 제공하여, 정책 관련 질문에 정확히 답변할 수 있도록 할 것입니다.

**Trustworthiness (신뢰성)**

Trustworthiness는 에이전트의 행동이 얼마나 믿을 수 있고 안전한지를 보장하는 개념입니다. 특히 에이전트가 실제 시스템에 영향을 미치는 도구(예: 쿠폰 발급, 주문 변경)를 사용할 때 매우 중요합니다. 신뢰성을 확보하기 위해 우리는 도구 사용 전에 특정 조건이 충족되는지 확인하는 로직을 추가할 수 있습니다. 예를 들어, 쿠폰 발급 도구는 '배송 지연' 상태인 주문에만 사용하도록 조건을 걸어, 에이전트가 실수로 정상적인 주문에 쿠폰을 발급하는 것을 막는 것입니다. 이는 에이전트가 예측 가능하고 안전하게 작동하도록 만듭니다.

**Domain-specific agent (도메인 특화 에이전트)**

Domain-specific agent는 특정 분야나 영역(도메인)에 대해 깊이 있는 지식과 기능을 갖춘 에이전트입니다. 예를 들어, 의료 분야 에이전트는 질병 진단이나 처방에 대한 전문 지식을 가지고 있고, 금융 분야 에이전트는 주식 거래나 투자 조언에 특화되어 있습니다. 우리 프로젝트에서 만드는 고객 서비스 에이전트는 'AI 온라인 서점'이라는 도메인에 특화되어, 주문, 배송, 보상, 쿠폰 등 고객 서비스와 관련된 질문과 요청을 처리하는 데 필요한 지식과 도구를 갖추게 됩니다. 특정 도메인에 특화되면 해당 분야의 복잡한 문제도 더 정확하고 효율적으로 해결할 수 있습니다.

### 1-1 정책 검색 도구 확장 (RAG)

이제 우리 에이전트의 지식 기반을 확장해 보겠습니다. Chapter 4-1 에서 배송 정책(`shipping_policy.txt`)으로 RAG를 구축했다면, 이번에는 보상 정책(`reward_policy.txt`)을 추가하여 에이전트가 배송뿐만 아니라 보상 및 보상 정책에 대해서도 정확하게 답변할 수 있도록 만들 것입니다.

이를 위해 먼저 `reward_policy.txt` 파일을 생성하고, 그 내용을 로드하여 기존 배송 정책과 함께 벡터 스토어에 저장하는 과정을 거칩니다.

#### `reward_policy.txt` 파일 준비

YES24의 정책에 관한 PDF 문서에서 보상 정책만을 텍스트로 추출합니다.

```python
import glob # 파일을 검색하기 위한 glob 모듈을 가져옵니다.
from langchain_community.document_loaders import TextLoader
from langchain_community.document_loaders import PyMuPDFLoader

# './' 경로에 있는 모든 PDF 파일을 찾습니다.
pdf_files = glob.glob('./*.pdf')

all_documents = [] # 모든 PDF 파일에서 로드된 문서를 저장할 빈 리스트를 생성합니다.

print(f"--- './' 경로에서 찾은 PDF 파일 목록: {pdf_files} ---")

# 찾은 각 PDF 파일에 대해 로드 및 파싱을 수행합니다.
for pdf_filepath in pdf_files:
    try:
        #loader = PyPDFLoader(pdf_filepath)
        loader = PyMuPDFLoader(pdf_filepath)
        pages = loader.load()
        all_documents.extend(pages) # 로드된 페이지들을 all_documents 리스트에 추가합니다.
        print(f"'{pdf_filepath}' 파일 로드 완료. {len(pages)} 페이지 로드됨.")
    except Exception as e:
        print(f"'{pdf_filepath}' 파일 로드 중 오류 발생: {e}")

# shipping_policy.txt 파일도 로드하여 all_documents에 추가합니다.
shipping_policy_file_path = "shipping_policy.txt"

try:
    # Create a TextLoader instance with the file path
    loader = TextLoader(shipping_policy_file_path)

    # Load the document
    text_documents = loader.load()

    # Extend all_documents with text file documents
    all_documents.extend(text_documents)
    print(f"'{shipping_policy_file_path}' 파일 로드 완료. {len(text_documents)} 문서 로드됨.")


except FileNotFoundError:
    print(f"Error: File not found at {shipping_policy_file_path}. Please ensure the file exists at this location.")
except Exception as e:
    print(f"An error occurred while loading {shipping_policy_file_path}: {e}")


# 모든 로드된 총 문서(PDF 페이지 + Text 문서) 정보 출력
print("\n--- 모든 파일에서 로드된 총 문서 정보 ---")
print(f"총 문서 개수: {len(all_documents)}")
if all_documents:
    print(f"첫 번째 문서 내용 (일부): {all_documents[0].page_content[:200]}...")
    print(f"첫 번째 문서 메타데이터: {all_documents[0].metadata}")
else:
    print("로드된 문서가 없습니다.")
print("-" * 20)

# 다음 단계를 위해 변수 이름을 'documents'로 맞춥니다.
documents = all_documents
```

```python
# 필요한 라이브러리를 임포트합니다.
from langchain_upstage import ChatUpstage
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter # 필요하다면 문서 분할에 사용
from langchain_core.prompts import ChatPromptTemplate

# Upstage LLM 인스턴스를 생성합니다.
# API 키는 환경 변수 UPSTAGE_API_KEY에 설정되어 있어야 합니다.
# 모델 이름은 필요에 따라 변경하세요 (예: "solar-1-mini-128k-instruct")
llm_upstage = ChatUpstage()

# LLM에게 보상 정책을 추출하도록 요청하는 프롬프트를 구성합니다.
# 모든 문서 내용을 하나의 문자열로 합치거나, 각 문서별로 LLM을 호출할 수 있습니다.
# 여기서는 간단하게 문서 내용을 합쳐서 프롬프트에 넣습니다.
all_document_content = "\n---\n".join([doc.page_content for doc in documents])

# LLM 프롬프트 정의
# LLM에게 문서 내용에서 보상 정책만 명확하게 추출하도록 지시합니다.
extraction_prompt_template = """다음은 다양한 정책 내용이 포함된 문서입니다.
이 문서에서 '보상 정책'과 관련된 내용은 모두 추출하여 정리해 주세요.
추출된 내용 외의 다른 설명이나 서론/결론은 포함하지 마세요.
보상 절차, 보상 조건, 보상 예외 사항 등 보상과 관련된 모든 정보를 포함해야 합니다.

문서 내용:
---
{document_content}
---

추출된 보상 정책:
"""

# 프롬프트에 문서 내용을 채웁니다.
extraction_prompt = extraction_prompt_template.format(document_content=all_document_content)
print('프롬프트 확인')
print("\n--- Upstage LLM을 사용하여 보상 정책 추출 시작 ---")

extracted_reward_policy = ""
try:
    # LLM 호출하여 보상 정책 추출
    # 긴 문서 내용의 경우, 청킹하여 각 청크별로 추출하거나 요약하는 방식이 더 효율적일 수 있습니다.
    # 여기서는 간단하게 전체 내용을 전달합니다 (모델의 컨텍스트 창 크기 고려 필요).
    llm_response = llm_upstage.invoke(extraction_prompt)
    print(llm_response.content )
    extracted_reward_policy = llm_response.content# 응답 객체에서 내용 추출

    print("보상 정책 추출 완료 (Upstage LLM). 추출된 내용 일부:")
    print(extracted_reward_policy[:500] + "...") # 내용 일부 출력
    print("----------------------------------------------------")

except Exception as e:
    print(f"Upstage LLM 호출 중 오류 발생: {e}")
    extracted_reward_policy = "보상 정책 추출 실패."


# --- 추출된 내용을 `reward_policy.txt` 파일로 저장 ---

# 추출된 보상 정책 내용을 `reward_policy.txt` 파일에 씁니다.
reward_policy_file_path = "reward_policy.txt"
if extracted_reward_policy and extracted_reward_policy != "보상 정책 추출 실패.":
    with open(reward_policy_file_path, "w", encoding='utf-8') as f: # 인코딩 지정
        f.write(extracted_reward_policy)
    print(f"`{reward_policy_file_path}` 파일이 추출된 보상 정책 내용으로 생성되었습니다.")
else:
    print(f"추출된 보상 정책 내용이 없거나 추출에 실패하여 `{reward_policy_file_path}` 파일을 생성하지 않았습니다.")


# --- `reward_policy.txt` 파일 로드 및 내용 확인 ---

# reward_policy.txt 파일 로드
try:
    # Create a TextLoader instance with the file path
    loader = TextLoader(reward_policy_file_path, encoding='utf-8') # 인코딩 지정

    # Load the document
    reward_docs = loader.load()
    print(f"\n'{reward_policy_file_path}' 파일 로드 완료. {len(reward_docs)} 문서 로드됨.")

    if reward_docs:
        print("--- 로드된 보상 정책 내용 확인 ---")
        print(reward_docs[0].page_content)
        print("-----------------------------")
    else:
        print("로딩된 보상 정책 문서가 비어 있습니다.")

except FileNotFoundError:
    print(f"Error: File not found at {reward_policy_file_path}. Please ensure the file exists after extraction.")
except Exception as e:
    print(f"An error occurred while loading or processing {reward_policy_file_path}: {e}")
```

#### Vector Store 확장 및 검색 확인

이제 준비된 `reward_policy.txt` 파일과 기존 `shipping_policy.txt` 파일을 함께 로드하여 RAG의 핵심인 Vector Store를 확장합니다. Vector Store는 문서를 임베딩(Embeddings, 벡터 변환)하여 저장해두는 곳으로, 검색 시 사용자의 질문을 벡터로 변환한 후 이 저장소에서 가장 유사한 문서 벡터를 찾아냅니다.

여기서는 LangChain의 `TextLoader`로 텍스트 파일을 읽고, `CharacterTextSplitter`로 문서를 적절한 크기의 '청크(Chunk)'로 분할하며, Mock 임베딩과 Mock Vector Store를 사용하여 벡터 스토어를 구축하고 검색 기능을 확인합니다. 실제 프로젝트에서는 `OpenAIEmbeddings`와 `Chroma` 같은 실제 라이브러리를 사용하게 됩니다.

**참고:**

*   **파싱(Parsing):** 비정형 데이터를 구조화된 형태로 변환하는 과정입니다. 여기서는 텍스트 파일의 내용을 읽어오는 것을 포함합니다.
*   **청킹(Chunking):** 긴 문서를 의미 있는 작은 단위(청크)로 나누는 과정입니다. RAG에서는 검색 및 임베딩 효율성을 위해 문서를 청크로 나눕니다. `CharacterTextSplitter`는 문자를 기준으로 청크를 나눕니다.
*   **임베딩(Embedding):** 텍스트를 컴퓨터가 이해할 수 있는 숫자 벡터로 변환하는 과정입니다. 의미적으로 유사한 텍스트는 유사한 벡터 값을 가집니다. `OpenAIEmbeddings`와 같은 임베딩 모델이 이 역할을 합니다.
*   **인덱싱(Indexing):** 검색을 빠르게 할 수 있도록 데이터를 정리하고 색인을 만드는 과정입니다. Vector Store에 문서를 저장할 때 내부적으로 인덱싱이 수행됩니다.
*   **검색기(Retriever):** 사용자의 질문과 가장 관련성이 높은 문서 청크를 Vector Store에서 찾아주는 역할을 합니다.

```python
# 필요한 라이브러리를 임포트합니다.
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
# 실제 프로젝트에서는 아래와 같은 임베딩 및 Vector Store를 사용합니다.
# from langchain_community.embeddings import OpenAIEmbeddings
# from langchain_community.vectorstores import Chroma
from langchain.agents import Tool # 에이전트 도구 정의를 위해 필요합니다.

# Chapter 4-1에서 사용했다고 가정한 get_order_status 함수를 정의합니다.
# 실제 주문 관리 시스템과 연동되는 대신, 간단한 더미 데이터를 반환합니다.
def get_order_status(order_id: str):
    """
    주문 ID를 받아 현재 배송 상태를 반환하는 도구입니다.
    실제 애플리케이션에서는 외부 주문 관리 시스템에 쿼리할 것입니다.
    """
    # 시연을 위해 특정 주문 ID에 대해 미리 정해진 상태를 반환합니다.
    if order_id == "ORDER123":
        return "Delivered" # 배송 완료
    elif order_id == "ORDER456":
        return "Shipping Delayed" # 배송 지연
    elif order_id == "ORDER789":
        return "Processing" # 처리 중
    else:
        return "Order Not Found" # 주문 찾을 수 없음

    print(f"모의: 주문 ID {order_id} 상태 확인 중")
    # 실제 시나리오에서는 외부 API를 호출합니다.
    # status = external_api.get_status(order_id)
    # return status


# 2. RAG Vector Store 확장 및 검색 확인
# `shipping_policy.txt` 파일 내용 (Chapter 4-1 에서 생성했다고 가정)
shipping_policy_content = """
배송 정책:
주문은 일반적으로 영업일 기준 1-2일 이내에 처리됩니다.
표준 배송은 영업일 기준 5-7일 소요됩니다.
빠른 배송은 영업일 기준 2-3일 소요됩니다.
일부 국가로는 국제 배송을 지원합니다.
예기치 못한 상황으로 인해 배송 지연이 발생할 수 있습니다.
"""
# `shipping_policy.txt` 파일 생성 (실행 시마다 존재하도록)
with open("shipping_policy.txt", "w") as f:
    f.write(shipping_policy_content)
print("`shipping_policy.txt` 파일이 생성되었습니다 (모의).")


# 문서 로드
# TextLoader를 사용하여 텍스트 파일의 내용을 불러옵니다.
shipping_loader = TextLoader("shipping_policy.txt")
reward_loader = TextLoader("reward_policy.txt")

shipping_docs = shipping_loader.load()
reward_docs = reward_loader.load()

# 두 파일의 문서를 하나로 합칩니다.
all_docs = shipping_docs + reward_docs

# 문서를 청크 단위로 분할
# CharacterTextSplitter를 사용하여 문자를 기준으로 청크를 나눕니다.
# chunk_size: 각 청크의 최대 문자 수
# chunk_overlap: 인접한 청크 간에 겹치는 문자 수 (검색 시 문맥 유지를 위해 사용)
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
split_docs = text_splitter.split_documents(all_docs)

print(f"총 {len(all_docs)}개의 문서를 로드했으며, {len(split_docs)}개의 청크로 분할했습니다.")

# 임베딩 생성 (시연을 위한 Mock 임베딩 사용)
# 실제 프로젝트에서는 OpenAIEmbeddings 등 실제 임베딩 모델을 사용합니다.
class MockEmbeddings:
    # 문서 목록을 받아 각 문서에 대한 임베딩 벡터 목록을 반환합니다.
    def embed_documents(self, texts):
        # 시연을 위해 모든 문서에 대해 동일한 더미 벡터를 반환합니다.
        print(f"모의: {len(texts)}개 문서 임베딩 중...")
        return [[1.0] * 10 for _ in texts] # 10차원 더미 벡터

    # 단일 쿼리 문자열을 받아 해당 쿼리에 대한 임베딩 벡터를 반환합니다.
    def embed_query(self, text):
        # 시연을 위해 쿼리에 대해 동일한 더미 벡터를 반환합니다.
        print(f"모의: 쿼리 '{text}' 임베딩 중...")
        return [1.0] * 10 # 10차원 더미 벡터

embeddings = MockEmbeddings()
print("Mock 임베딩 모델이 준비되었습니다.")


# Vector Store 생성 (시연을 위한 Mock Vector Store 사용)
# 실제 프로젝트에서는 Chroma.from_documents 등 실제 Vector Store를 사용합니다.
# Vector Store는 임베딩된 문서 청크를 저장하고 검색하는 기능을 제공합니다.
class MockVectorStore:
    def __init__(self, docs, embeddings):
        self.docs = docs # 저장된 문서 청크 목록
        self.embeddings = embeddings # 사용할 임베딩 모델
        # 실제 Vector Store는 여기서 임베딩을 수행하고 인덱스를 구축합니다.
        # self.index = self.embeddings.embed_documents([doc.page_content for doc in docs])
        print(f"모의: {len(self.docs)}개 문서로 Vector Store 초기화 완료.")


    # Vector Store를 검색기(Retriever) 객체로 변환하여 반환합니다.
    def as_retriever(self):
        # 실제 구현에서는 검색 설정이 적용된 Retriever 객체를 반환합니다.
        # 이 모의 객체는 get_relevant_documents 메서드를 직접 제공하여 검색기 역할을 대신합니다.
        return self

    # 쿼리 문자열을 받아 가장 관련성이 높은 문서 목록을 반환합니다.
    def get_relevant_documents(self, query):
        print(f"모의 RAG 검색 요청: '{query}'")
        # 실제 검색 로직 대신, 쿼리 키워드를 기반으로 관련 문서가 있는 척 시뮬레이션합니다.
        query_lower = query.lower()
        results = []
        for doc in self.docs:
            # 쿼리에 'shipping' 또는 '배송'이 포함되고 문서 내용에 'shipping' 또는 '배송'이 있으면 추가
            if ('shipping' in query_lower or '배송' in query_lower) and ('shipping' in doc.page_content.lower() or '배송' in doc.page_content.lower()):
                 results.append(doc)
            # 쿼리에 'reward' 또는 '보상'이 포함되고 문서 내용에 'reward' 또는 '보상'이 있으면 추가
            elif ('reward' in query_lower or '보상' in query_lower) and ('reward' in doc.page_content.lower() or '보상' in doc.page_content.lower()):
                 results.append(doc)
            # 간단한 휴리스틱: 쿼리 단어가 문서 내용에 포함되어 있으면 추가 (정확도는 낮음)
            elif any(word in doc.page_content.lower() for word in query_lower.split()):
                 results.append(doc)

        # 중복 결과를 피하고 결과 수를 제한합니다.
        unique_results = []
        seen_content = set()
        for r in results:
            if r.page_content not in seen_content:
                unique_results.append(r)
                seen_content.add(r.page_content)

        print(f"모의 RAG 검색 결과: {len(unique_results)}개 문서 발견 (최대 3개 반환)")
        return unique_results[:3] # 최대 3개의 관련 문서 반환


vectorstore = MockVectorStore(split_docs, embeddings)
retriever = vectorstore.as_retriever()# Vector Store에서 검색기를 얻습니다.

print("RAG Vector Store가 배송 및 보상 정책으로 확장되었습니다.")

# 3. 확장된 RAG 검색 기능 확인
print("\n--- RAG 검색 기능 확인 ---")

# 보상 정책 관련 질문으로 검색 테스트
reward_query = "보상 정책이 어떻게 되나요?"
print(f"\n'{reward_query}' 검색 결과:")
reward_results = retriever.get_relevant_documents(reward_query)
for i, doc in enumerate(reward_results):
    print(f"문서 {i+1}:\n{doc.page_content}\n---")

# 배송 정책 관련 질문으로 검색 테스트
shipping_query = "배송은 얼마나 걸리나요?"
print(f"\n'{shipping_query}' 검색 결과:")
shipping_results = retriever.get_relevant_documents(shipping_query)
for i, doc in enumerate(shipping_results):
    print(f"문서 {i+1}:\n{doc.page_content}\n---")

print("--- RAG 검색 기능 확인 완료 ---")
```

### 1-2 불만 고객 쿠폰 발급 도구 개발 및 Trustworthiness 구현

이제 고객의 불만 사항에 대해 보상을 제공할 수 있는 새로운 도구를 만듭니다. 바로 '불만 고객 쿠폰 발급 도구'입니다. 이 도구는 특정 주문 ID에 대해 할인 쿠폰을 발급하는 역할을 합니다.

여기서 중요한 것은 '신뢰성(Trustworthiness)' 개념을 적용하는 것입니다. 우리는 에이전트가 아무 주문에나 쿠폰을 남발하는 것을 막아야 합니다. 따라서 이 도구는 쿠폰을 발급하기 전에 반드시 해당 주문의 상태를 확인하고, '배송 지연'과 같이 쿠폰 발급 조건에 맞는 경우에만 실제로 쿠폰을 발급하도록 로직을 추가할 것입니다. `get_order_status` 함수를 활용하여 주문 상태를 확인하는 것이 신뢰성 구현의 핵심입니다.

```python
# 4. 불만 고객 쿠폰 발급 도구 개발 및 Trustworthiness 구현

# issue_complaint_coupon 함수를 정의합니다.
# 이 함수는 주문 ID를 입력받아 쿠폰 발급 로직을 수행합니다.
def issue_complaint_coupon(order_id: str):
    """
    지정된 주문 ID에 대해 불만 고객 쿠폰을 발급하는 도구입니다.
    Trustworthiness 로직이 포함되어 있어 주문 상태가 'Shipping Delayed'일 때만 쿠폰을 발급합니다.
    """
    print(f"쿠폰 발급 시도 중: 주문 ID {order_id}")

    # Trustworthiness 구현: get_order_status 도구를 사용하여 주문 상태를 확인합니다.
    order_status = get_order_status(order_id) # 이전에 정의한 get_order_status 함수 호출

    # 주문 상태가 'Shipping Delayed'인지 확인합니다.
    if order_status == "Shipping Delayed":
        # 상태가 '배송 지연'이면 쿠폰 발급 메시지를 반환합니다.
        # 실제 시나리오에서는 외부 시스템에 쿠폰 발급 요청을 보낼 것입니다.
        print(f"모의: 지연된 주문 ID {order_id}에 대해 쿠폰 발급 완료.")
        return f"주문 ID {order_id}에 대해 5,000원 할인 쿠폰이 발급되었습니다."
    else:
        # 그 외의 상태이면 쿠폰 발급 대상이 아님을 알립니다.
        print(f"모의: 주문 ID {order_id} 상태 '{order_status}'는 쿠폰 발급 대상이 아닙니다.")
        return f"주문 ID {order_id}는 쿠폰 발급 대상이 아닙니다. 현재 상태: {order_status}"

print("`issue_complaint_coupon` 도구 함수 개발 완료 (Trustworthiness 로직 포함).")

# 개발된 도구 테스트 (Trustworthiness 로직 확인)
print("\n--- 쿠폰 발급 도구 테스트 ---")
# 배송 지연 상태인 주문 테스트
test_order_delayed = "ORDER456"
coupon_result_delayed = issue_complaint_coupon(test_order_delayed)
print(f"주문 ID {test_order_delayed} 쿠폰 발급 결과: {coupon_result_delayed}")

# 배송 완료 상태인 주문 테스트
test_order_delivered = "ORDER123"
coupon_result_delivered = issue_complaint_coupon(test_order_delivered)
print(f"주문 ID {test_order_delivered} 쿠폰 발급 결과: {coupon_result_delivered}")

# 존재하지 않는 주문 테스트
test_order_unknown = "ORDER_UNKNOWN"
coupon_result_unknown = issue_complaint_coupon(test_order_unknown)
print(f"주문 ID {test_order_unknown} 쿠폰 발급 결과: {coupon_result_unknown}")
print("--- 쿠폰 발급 도구 테스트 완료 ---")
```

## 생각의 흐름 만들기 (Planning & Reasoning)

현재 Agent는 질문을 받으면 바로 도구를 사용하려고 할 수 있습니다. 하지만 복잡한 민원(예: 배송 지연 불만 및 보상 요구)을 해결하려면 단순히 도구를 사용하는 것을 넘어, 문제 상황을 분석하고, 해결을 위한 단계를 스스로 생각하며(Reasoning), 어떤 도구를 어떤 순서로 사용할지 계획(Planning)을 세우는 과정이 필요합니다.

이번 과제에서는 LangChain의 ReAct(Reason+Act) 프레임워크를 적용하여 Agent에게 '생각하는 능력'을 부여하고, 문제 해결을 위한 논리적인 흐름을 따르도록 가이드하는 방법을 배웁니다. ReAct는 생각(Thought)과 행동(Action)을 번갈아 수행하며 목표를 달성하는 방식입니다.

### 학습 키워드 설명

**Reasoning (추론)**

Reasoning은 에이전트가 주어진 정보(사용자 요청, 도구 결과 등)를 바탕으로 논리적인 판단을 내리고 결론을 도출하는 과정입니다. 고객의 요청의 진짜 의도가 무엇인지 파악하고, 어떤 정보가 부족하며, 다음 단계로 무엇을 해야 할지 스스로 생각하는 능력입니다. ReAct 프레임워크에서 'Thought' 부분이 바로 이 추론 과정에 해당합니다.

**Planning (계획 수립)**

Planning은 에이전트가 추론을 통해 파악한 목표를 달성하기 위해 구체적인 행동 순서, 즉 계획을 세우는 과정입니다. 어떤 도구를 먼저 사용하고, 그 결과에 따라 다음 도구를 어떻게 사용할지 등 일련의 절차를 설계합니다. 복잡한 문제를 해결하기 위해서는 효과적인 계획 수립이 필수적입니다.

**Action (행동)**

Action은 계획에 따라 에이전트가 실제로 특정 도구를 사용하거나 외부 시스템과 상호작용하는 실행 단계입니다. 예를 들어, '주문 상태 확인' 도구를 사용하거나 '쿠폰 발급' 도구를 실행하는 것이 Action입니다. Action을 실행한 결과는 'Observation'(관찰 결과)으로 에이전트에게 돌아오며, 에이전트는 이 Observation을 바탕으로 다음 Thought와 Action을 결정합니다.

**ReAct (Reasoning and Acting, 추론 및 행동)**

ReAct는 Reasoning(추론)과 Acting(행동)을 반복적으로 수행하며 목표를 달성하는 에이전트 프레임워크입니다. 에이전트는 사용자 요청을 받으면 먼저 **Thought**(생각)를 통해 상황을 분석하고 다음에 할 행동을 결정합니다. 그리고 결정한 **Action**(행동)을 실행합니다. Action의 결과로 **Observation**(관찰)을 얻게 되고, 이 Observation을 바탕으로 다시 **Thought**를 하여 다음 Action을 결정합니다. 이 과정을 목표 달성 시까지 반복하여, 동적으로 문제를 해결해 나갑니다. 이는 에이전트가 미리 정해진 규칙만 따르는 것이 아니라, 상황에 맞춰 유연하게 대처할 수 있게 해줍니다.

ReAct 흐름 요약:
사용자 요청 -> 계획수립 -> Thought -> Action -> Observation -> Thought -> Action -> Observation -> ... -> Thought -> Final Answer

### 2-1 ReAct 프레임워크 적용을 위한 시스템 프롬프트 수정

Agent가 ReAct 프레임워크에 따라 작동하도록 만들려면, Agent에게 주어지는 첫 번째 지시문인 '시스템 프롬프트(System Prompt)'를 수정해야 합니다. 이 프롬프트는 Agent의 역할, 사용할 수 있는 도구, 그리고 문제를 해결하기 위해 따라야 할 절차를 정의합니다.

우리는 시스템 프롬프트에 Agent가 Thinking(Thought), Acting(Action), Observing(Observation) 단계를 반복하도록 명확하게 지시하는 내용을 추가할 것입니다. 또한, Agent가 스스로의 사고 과정을 설명하도록 유도하여, 우리가 Agent의 결정 과정을 이해하고 디버깅하는 데 도움을 줄 것입니다.

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import AIMessage, HumanMessage

# ReAct 에이전트가 사용할 시스템 프롬프트를 정의합니다.
# 이 프롬프트는 Agent의 역할, 사용 가능한 도구 목록, 그리고 ReAct의 Thought/Action/Observation
# 단계를 따르도록 하는 지시문을 포함합니다.

# {tools}: 사용 가능한 도구 목록과 설명이 create_react_agent에 의해 자동으로 채워집니다.
# {agent_scratchpad}: 이전 Thought, Action, Observation 기록이 AgentExecutor에 의해 자동으로 채워집니다.

system_prompt = """당신은 친절하고 유능한 고객 서비스 에이전트입니다.
당신은 고객의 요청을 처리하기 위해 다음 도구들에 접근할 수 있습니다:

{tools}

고객을 돕기 위해 ReAct 프레임워크를 사용하여 추론(Thought), 행동(Action), 관찰 결과(Observation) 단계를 번갈아 수행해야 합니다.

반드시 다음 형식으로 응답해야 합니다:

Thought: 지금 생각하고 있는 것을 설명합니다.
Action: 사용할 도구 이름(사용할 도구의 입력값)
Observation: 도구 실행 결과 (이것은 시스템에 의해 채워집니다)
... (이 Thought/Action/Observation 과정을 필요에 따라 반복합니다)
Thought: 모든 정보가 충분하고 최종 답변을 제공할 준비가 되었음을 인식합니다.
Final Answer: 최종 답변입니다. (고객에게 제공할 응답)

사용할 수 있는 도구 이름은 {tool_names} 입니다.

따라야 할 절차는 다음과 같습니다:
1.  **Thought:** 사용자의 요청과 사용 가능한 도구들을 신중하게 분석합니다. 사용자의 요청을 해결하기 위해 다음에 수행할 최선의 단계를 결정합니다. 추론 과정을 명확하게 설명하세요.
2.  **Action:** 목록에서 가장 적절한 도구를 선택하고 해당 도구에 필요한 입력을 제공합니다. 형식은 `도구 이름[도구 입력값]` 과 같아야 합니다. Action 다음에는 반드시 Observation이 따라옵니다.
3.  **Observation:** Agent가 실행한 Action의 결과입니다. 이 결과를 바탕으로 다음 Thought를 진행합니다. (이 부분은 시스템에 의해 제공됩니다)
4.  Thought/Action/Observation 과정을 반복하여 문제를 해결합니다.
5.  문제를 해결했거나 더 이상 진행할 단계가 없으면, 최종 Thought를 설명하고 **Final Answer:** 형식으로 최종 답변을 제공합니다.

이전 대화 기록과 현재 요청을 고려하여 응답하세요.
"""

# ReAct 에이전트 프롬프트 템플릿을 생성합니다.
# 시스템 프롬프트와 함께 사용자 입력({input}) 및 Agent 중간 단계({agent_scratchpad})를 포함합니다.
prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    ("human", "{input}\n\n{agent_scratchpad}"),
])

print("ReAct 에이전트 시스템 프롬프트 및 템플릿이 정의되었습니다.")
```

### 2-2 Agent 실행 및 '생각의 과정(Thought)' 로그 확인 방법 안내

이제 ReAct 프레임워크 지시문이 포함된 시스템 프롬프트와 우리가 만든 도구들을 사용하여 Agent 실행 체인을 구성하고 실제로 Agent를 실행해 보겠습니다. Agent가 복잡한 질문을 받았을 때, 이전에 정의한 ReAct 절차에 따라 어떻게 '생각(Thought)'하고 '행동(Action)'하며 '관찰(Observation)'하는지 그 과정을 로그를 통해 확인할 수 있습니다.

여기서는 실제 Language Model(LLM) 대신 시뮬레이션을 위한 Mock LLM을 사용합니다. 이 Mock LLM은 특정 입력에 대해 미리 정해진 ReAct 응답을 반환하도록 설정되어 있어, 실제 LLM API 키 없이도 Agent의 ReAct 흐름을 테스트하고 로그를 확인하는 방법을 보여줄 수 있습니다.

Agent 실행 시 `verbose=True` 옵션을 설정하면, LangChain이 Agent의 내부 작업 과정, 즉 Thought, Action, Observation 단계를 상세하게 콘솔에 출력해 줍니다.

## 다른 팀에게 업무 요청하기 (MCP & Multi-Agent), 메모리 붙이기

CS Agent는 고객의 요청을 직접 처리하는 것 외에도, 다른 시스템이나 에이전트와 협력해야 할 때가 있습니다. 예를 들어, 고객에게 보상을 제공하기로 결정했다면, 실제 쿠폰을 발급하거나 보상을 처리하는 시스템에 이 정보를 전달해야 합니다. 또한, 고객과의 대화가 여러 번 오가더라도 이전 내용을 기억하고 맥락을 유지하는 것이 중요합니다.

이번 과제에서는 다음 두 가지 주요 목표를 달성합니다:

1.  **다른 시스템과의 통신 준비 (MCP & JSON 출력):** Agent가 다른 시스템(예: 결제 시스템)이 이해할 수 있는 표준화된 형식(JSON)으로 '업무 요청서'를 생성하도록 만듭니다. 이를 통해 시스템 간 메시지 통신 규약(MCP)의 중요성을 배우고, Pydantic 모델과 출력 파서를 사용하여 JSON 출력을 강제하는 방법을 실습합니다.
2.  **멀티턴 대화 기능 추가 (메모리):** Agent가 이전 대화 내용을 기억하여 사용자와의 상호작용에서 연속성을 갖도록 메모리 기능을 Agent에 통합합니다.

### 학습 키워드 설명

**MCP (Model Context Protocol, 모델 문맥 프로토콜)**

MCP는 다양한 시스템, 서비스 또는 에이전트 간에 정보를 교환하기 위한 표준화된 데이터 형식 및 통신 규칙의 집합입니다. 에이전트가 다른 시스템에 특정 작업을 요청하거나 정보를 전달할 때, 미리 약속된 형식(예: JSON)으로 메시지를 주고받으면 통신 오류를 줄이고 시스템 간의 통합을 용이하게 할 수 있습니다. 마치 사람들이 대화할 때 사용하는 '언어'처럼, 컴퓨터 시스템들은 MCP라는 '표준 데이터 언어'를 사용하여 서로를 이해하고 협력합니다. 우리 프로젝트에서는 Agent가 결제 시스템에 보상 처리를 요청할 때 사용할 JSON 형식을 정의하고, Agent의 출력이 이 형식을 따르도록 만드는 과정을 통해 MCP의 개념을 이해합니다.

**Multi-agent interaction (멀티 에이전트 상호작용)**

Multi-agent interaction은 여러 개의 독립적인 AI 에이전트가 서로 정보를 교환하고 협력하여 공동의 목표를 달성하거나 복잡한 문제를 해결하는 시나리오를 말합니다. 각 에이전트는 고유한 역할, 능력, 정보를 가질 수 있으며, 서로의 강점을 활용하여 혼자서는 해결하기 어려운 과제를 수행합니다. 예를 들어, CS Agent는 고객 상담을 전문으로 하고, Billing Agent는 결제 및 보상 처리를 전문으로 하며, 둘이 MCP를 통해 소통하여 고객의 보상 요청을 완벽하게 처리할 수 있습니다. 이번 과제에서 Agent가 JSON 업무 요청서를 생성하는 것은 멀티 에이전트 시스템의 기초적인 상호작용 형태를 시뮬레이션하는 것입니다.

**Memory (메모리)**

AI 에이전트에서 메모리는 이전 대화 내용을 저장하고 관리하는 기능입니다. 사람이 대화할 때 앞서 나눈 이야기를 기억하고 그 맥락 위에서 다음 대화를 이어가듯이, 에이전트도 메모리를 통해 사용자와의 과거 상호작용 기록(질문, 답변, 도구 사용 결과 등)을 참고하여 현재의 요청을 더 정확하게 이해하고 적절히 응답할 수 있습니다. 메모리가 없으면 에이전트는 매번 새로운 질문을 받을 때마다 이전의 모든 내용을 잊어버리고 처음부터 다시 시작해야 하므로 자연스럽고 효율적인 대화가 어렵습니다. LangChain에서는 `ConversationBufferMemory`와 같은 클래스를 제공하여 이러한 메모리 기능을 쉽게 구현할 수 있습니다.

```python
# 1. Define the BillingRequest Pydantic model
from pydantic import BaseModel, Field, ValidationError # ValidationError 임포트 추가
import json # json 임포트 추가

# Pydantic BaseModel을 상속받아 BillingRequest 데이터 모델을 정의합니다.
# 이 모델은 다른 시스템(예: 결제 시스템)에 보상 요청 정보를 전달할 때 사용할 데이터의 구조를 정의합니다.
class BillingRequest(BaseModel):
    """
    결제 시스템에 보상 처리를 요청하기 위한 데이터 모델입니다.
    """
    # order_id: 보상이 요청된 주문의 ID (문자열, 필수)
    order_id: str = Field(..., description="보상이 요청된 주문의 ID.")
    # amount: 제안된 보상 금액 (실수, 필수)
    amount: float = Field(..., description= "제안된 보상 금액.")
    # reason: 보상 요청 사유 (문자열, 필수)
    reason: str = Field(..., description= "보상 요청 사유 (예: 배송 지연, 상품 손상).")

print("\nBillingRequest Pydantic 모델이 정의되었습니다.")
# 정의된 모델의 JSON 스키마를 출력하여 구조를 확인합니다.
print("BillingRequest JSON 스키마:")
print(BillingRequest.model_json_schema())
```

### 멀티턴 대화를 위한 Agent 메모리 기능 추가

Agent가 사용자와의 대화 맥락을 유지하려면 이전 대화 내용을 '기억'해야 합니다. LangChain에서는 다양한 메모리 타입을 제공하며, 그 중 `ConversationBufferMemory`는 대화 기록 전체를 버퍼에 저장하는 간단한 메모리 구현체입니다.

이 과제에서는 `ConversationBufferMemory`를 사용하여 Agent 실행 체인에 메모리를 추가하고, 멀티턴 대화 테스트를 통해 Agent가 이전 대화를 기억하는지 확인합니다.

```python
# 필요한 라이브러리를 임포트합니다.
from langchain.agents import create_react_agent, AgentExecutor
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import AIMessage, HumanMessage
from langchain_upstage import ChatUpstage
from langchain.memory import ConversationBufferMemory # 메모리 클래스 임포트
from langchain.agents import Tool # Tool 클래스 임포트 (이전 셀에서 정의되었다고 가정)


# Assume llm (ChatUpstage instance), agent_tools (list of Tools), and prompt (ChatPromptTemplate) are defined in previous cells.

# 메모리 인스턴스를 생성합니다.
# return_messages=True로 설정하면 대화 기록을 메시지 객체 리스트로 반환합니다.
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)

print("ConversationBufferMemory 인스턴스가 생성되었습니다.")

# 메모리가 추가된 Agent 실행 체인을 구성합니다.
# AgentExecutor를 생성할 때 memory 인자에 메모리 인스턴스를 전달합니다.
# 또한, Agent 프롬프트가 {chat_history} 변수를 포함하도록 수정되어야 합니다.
# (이전 ReAct 프롬프트는 이미 {chat_history}를 포함하고 있습니다.)
agent_executor_with_memory = AgentExecutor(
    agent=create_react_agent(llm, agent_tools, prompt), # 이전 llm, agent_tools, prompt 사용
    tools=agent_tools,
    verbose=True, # Agent의 상세 로그 출력
    memory=memory # 메모리 인스턴스 연결
)

print("\n메모리가 연결된 Agent 실행 체인 구성 완료.")


# 멀티턴 대화 테스트
print("\n## 멀티턴 대화 테스트 시작 ##")

# 첫 번째 턴: 주문 상태 확인 (이전 대화 맥락 없음)
first_query = "주문 ID ORDER123의 상태를 알려주세요."
print(f"\n--- 사용자: {first_query} ---")
try:
    result_1 = agent_executor_with_memory.invoke({"input": first_query})
    print(f"--- Agent 응답: {result_1['output']} ---")
except Exception as e:
    print(f"첫 번째 턴 실행 중 오류 발생: {e}")


# 두 번째 턴: 이전 주문에 대한 추가 질문 (주문 ID를 다시 언급하지 않음)
# Agent가 이전 턴의 'ORDER123'을 기억해야 합니다.
second_query = "그 주문이 배송 완료되면 알려줄 수 있나요?" # ORDER123에 대한 질문
print(f"\n--- 사용자: {second_query} ---")
try:
    # invoke 시 chat_history를 명시적으로 전달하지 않아도 AgentExecutor가 memory 객체를 사용합니다.
    # 하지만 ReAct 프롬프트가 chat_history를 받도록 설계되어 있다면
    # {"input": second_query, "chat_history": memory.load_memory_variables({}).get("chat_history")} 와 같이
    # memory의 내용을 명시적으로 전달해 줄 수도 있습니다. AgentExecutor는 memory 객체가 있으면 자동으로 전달해 줍니다.
    result_2 = agent_executor_with_memory.invoke({"input": second_query})
    print(f"--- Agent 응답: {result_2['output']} ---")
except Exception as e:
    print(f"두 번째 턴 실행 중 오류 발생: {e}")


print("\n## 멀티턴 대화 테스트 종료 ##")

# 메모리에 저장된 대화 기록 확인 (선택 사항)
print("\n--- 메모리에 저장된 대화 기록 ---")
print(memory.load_memory_variables({})["chat_history"])
print("---------------------------------")

# Note: 'llm', 'agent_tools', 'prompt' 변수가 이전 셀에서 정의되어 있어야 합니다.
# 'get_order_status', 'issue_complaint_coupon' 함수도 필요합니다.
```

### 학습 키워드 설명

**Reward Model (보상 모델)**

보상 모델은 강화 학습(Reinforcement Learning)에서 사용되는 개념으로, AI 에이전트의 행동이 얼마나 '좋은지' 또는 '바람직한지'를 평가하여 수치화된 보상(Reward)을 제공하는 모델입니다. 에이전트는 이 보상을 최대화하는 방향으로 학습하게 됩니다. 고객 서비스 에이전트의 경우, 고객 만족도, 문제 해결까지 걸린 시간, 정책 준수 여부, 그리고 이번 과제에서 다룰 '도구 사용의 논리적 순서' 등이 보상의 기준이 될 수 있습니다. 보상 모델은 에이전트가 복잡한 목표를 달성하기 위해 어떤 행동을 해야 하는지 학습하도록 안내하는 역할을 합니다.
프로젝트 2에서는 보상 모델을 직접 구축하고 학습시키지는 않지만, '도구 사용의 논리적 순서'를 평가하는 함수를 개선하는 것은 사실상 에이전트의 행동을 평가하는 '보상'의 한 형태를 정의하는 것과 같습니다. 이는 더 나아가 에이전트의 성능을 측정하고 개선하는 데 중요한 기반이 됩니다.

**Planning (계획 수립)**

계획 수립은 AI 에이전트가 주어진 목표를 달성하기 위해 일련의 행동 순서를 결정하는 과정입니다. 특히 복잡한 환경에서 여러 단계를 거쳐야 하거나 다양한 도구를 사용해야 할 때 중요합니다. 에이전트는 현재 상태, 목표, 사용 가능한 도구 및 그 기능에 대한 이해를 바탕으로 최적의 행동 경로를 탐색합니다. ReAct 프레임워크에서 'Thought' 단계는 계획 수립 과정의 일부로 볼 수 있습니다. 에이전트는 'Thought'를 통해 현재 상황을 분석하고, 어떤 도구를 사용해야 목표에 도달할 수 있을지 계획을 세웁니다.

```python
import json
from typing import List, Dict, Any

# Assume agent_tools and get_order_status are defined from previous steps
# Assume a mock_agent_executor that produces detailed ReAct logs is available or can be simulated

# 에이전트 실행을 시뮬레이션하고 로그를 캡처하는 플레이스홀더 함수입니다.
# 실제 시나리오에서는 AgentExecutor의 상세 출력을 캡처합니다.
def run_agent_and_capture_logs(agent_executor, query: str, chat_history: List[Dict[str, Any]]) -> List[str]:
    """
    에이전트 실행을 시뮬레이션하고 ReAct 단계를 나타내는 로그 라인 목록을 캡처합니다.
    (Thought, Action, Observation, Final Answer).
    실제 구현에서는 상세 출력을 파싱합니다.
    """
    print(f"\n--- 에이전트 실행 시뮬레이션 및 로그 캡처 중: '{query}' ---")
    # 특정 지연 배송 시나리오에 대한 모의 LLM의 출력 구조를 기반으로 한 단순화된 시뮬레이션입니다.
    logs = []
    simulated_steps = [
        "Thought: 사용자의 요청은 주문 상태를 확인하고, 배송이 지연된 경우 쿠폰 발급 가능성을 문의하는 것입니다. 이를 위해 먼저 주문 ID ORDER456의 현재 배송 상태를 확인해야 합니다. 'Get Order Status' 도구를 사용하여 주문 상태를 조회하겠습니다.",
        "Action: Get Order Status[ORDER456]",
        "Observation: 주문 ID ORDER456의 배송 상태는 'Shipping Delayed' 입니다.",
        "Thought: 주문 상태가 'Shipping Delayed'임을 확인했습니다. 이제 배송 지연에 대한 보상 정책을 확인하고, 해당 정책에 따라 쿠폰 발급이 가능한지 판단해야 합니다. (실제는 RAG 검색 필요)", # 시뮬레이션: Observation 후 Thought
        "Action: Policy Search[배송 지연 보상 정책]", # 시뮬레이션: RAG Action
        "Observation: 배송 지연 보상 정책: 배송 지연이 발생한 경우, 고객은 10% 할인 쿠폰을 받을 수 있습니다. (모의 RAG 결과)", # 시뮬레이션: RAG Observation
        "Thought: 배송 지연 보상 정책을 검토한 결과, 고객이 10% 할인 쿠폰을 받을 자격이 있음을 확인했습니다. 이제 불만 쿠폰을 발급해보겠습니다.", # 시뮬레이션: RAG 후 Thought
        "Action: Issue Complaint Coupon[ORDER456]",
        "Observation: 주문 ID ORDER456에 대해 5,000원 할인 쿠폰이 발급되었습니다.",
        "Thought: 모든 정보를 수집하고 문제를 해결했습니다. 고객에게 배송 상태를 알리고, 쿠폰이 발급되었음을 알려주면 됩니다.",
        "Final Answer: 고객님, 주문 ID ORDER456의 현재 배송 상태는 지연되었습니다. 불편을 드려 죄송합니다. 지연에 대한 보상으로 주문 ID ORDER456에 대해 5,000원 할인 쿠폰이 발급되었습니다. 추가 문의사항이 있으시면 언제든지 말씀해주세요."
    ]

    # 데모를 위해 쿼리에 따라 다른 시나리오를 시뮬레이션합니다.
    if "ORDER123" in query and "배송 상태" in query: # 배송 완료 시나리오
        simulated_steps = [
            "Thought: 사용자의 요청은 주문 상태를 확인하는 것입니다. 'Get Order Status' 도구를 사용하여 주문 ID ORDER123의 배송 상태를 조회하겠습니다.",
            "Action: Get Order Status[ORDER123]",
            "Observation: 주문 ID ORDER123의 배송 상태는 'Delivered' 입니다.",
            "Thought: 주문 상태를 확인했습니다. 사용자에게 배송 완료 상태를 안내합니다.",
            "Final Answer: 고객님, 주문 ID ORDER123의 현재 배송 상태는 'Delivered' 입니다."
        ]
    elif "ORDER789" in query and "상태 확인" in query: # 처리 중 시나리오
         simulated_steps = [
            "Thought: 사용자의 요청은 주문 상태를 확인하는 것입니다. 'Get Order Status' 도구를 사용하여 주문 ID ORDER789의 배송 상태를 조회하겠습니다.",
            "Action: Get Order Status[ORDER789]",
            "Observation: 주문 ID ORDER789의 배송 상태는 'Processing' 입니다.",
            "Thought: 주문 상태를 확인했습니다. 사용자에게 현재 처리 중인 상태를 안내합니다.",
            "Final Answer: 고객님, 주문 ID ORDER789의 현재 배송 상태는 'Processing' 입니다."
        ]
    elif "ORDER_UNKNOWN" in query and "상태 알려줘" in query: # 알 수 없는 주문 ID 시나리오
         simulated_steps = [
            "Thought: 사용자의 요청은 주문 상태를 확인하는 것입니다. 'Get Order Status' 도구를 사용하여 주문 ID ORDER_UNKNOWN의 배송 상태를 조회하겠습니다.",
            "Action: Get Order Status[ORDER_UNKNOWN]",
            "Observation: Order Not Found",
            "Thought: 주문 ID를 찾을 수 없습니다. 고객에게 주문 ID를 다시 확인해달라고 요청합니다.",
            "Final Answer: 고객님, 죄송하지만 주문 ID ORDER_UNKNOWN를 찾을 수 없습니다. 주문 ID를 다시 한 번 확인해 주시거나 다른 정보를 알려주시면 감사하겠습니다."
        ]
    elif "환불 정책" in query or "환불" in query: # 환불 정책 시나리오
        simulated_steps = [
            "Thought: 사용자의 요청은 환불 정책에 대한 정보입니다. 'Policy Search' 도구를 사용하여 관련 정책 문서를 검색해야 합니다.",
            "Action: Policy Search[환불 정책에 대해 알려줄 수 있나요?]", # 정책 검색 액션 시뮬레이션
            "Observation: 환불 정책: 고객은 구매일로부터 30일 이내에 제품이 사용되지 않았고 원래 포장 상태를 유지하는 경우 전액 환불을 요청할 수 있습니다...", # 모의 RAG 결과
            "Thought: 관련 정책 정보를 검색했습니다. 사용자에게 정책 내용을 요약하여 안내합니다.",
            "Final Answer: 저희 환불 정책에 따르면, 상품이 미사용 상태이고 원래 포장 그대로인 경우 구매일로부터 30일 이내에 전액 환불을 요청할 수 있습니다..."
        ]
    elif "쿠폰" in query and "발급" in query: # 쿠폰 문의 시나리오 (상태 확인이 먼저 필요할 수 있음)
        simulated_steps = [
            "Thought: 고객이 쿠폰 발급을 요청했습니다. 쿠폰 발급은 특정 조건(예: 배송 지연)을 만족할 때만 가능합니다. 요청에 주문 ID가 포함되어 있는지 확인하고, 주문 상태를 먼저 확인해야 합니다.",
            "Final Answer: 고객님, 쿠폰 발급은 배송 지연과 같은 특정 조건에 해당될 때 가능합니다. 어떤 주문에 대해 쿠폰을 문의하시는지 주문 ID를 알려주시거나 상황을 자세히 설명해 주시면 확인 후 도와드리겠습니다."
        ]
    elif chat_history and "배송 완료" in query and "알려줄 수 있나요" in query: # 멀티턴 후속 질문 (시뮬레이션)
        # 이전 대화 기록을 기반으로 시뮬레이션합니다. run_agent_and_capture_logs 함수는
        # 쿼리와 chat_history 확인을 기반으로 이를 시뮬레이션합니다.
        # 예: 사용자가 상태를 묻고(턴 1), "그 주문"의 알림에 대해 묻습니다(턴 2).
        # chat history를 구축하기 위해 턴 1을 먼저 시뮬레이션합니다 (개념적으로).
        # 그런 다음 시뮬레이션된 chat history를 전달하여 후속 쿼리로 턴 2를 시뮬레이션합니다.
        simulated_steps = [
            "Thought: 고객이 이전 대화에서 언급된 주문의 배송 완료 시 알림 가능성을 묻고 있습니다. 메모리를 통해 이전 대화 맥락에서 주문 ID를 파악하고, 배송 완료 알림 정책을 검색하여 안내해야 합니다.",
            "Action: Policy Search[배송 완료 알림 정책]", # 시뮬레이션: RAG Action
            "Observation: 배송 완료 알림 정책: 고객은 주문 관리 페이지에서 배송 알림을 설정할 수 있습니다. (모의 RAG 결과)", # 시뮬레이션: RAG Observation
            "Thought: 배송 완료 알림을 받을 수 있는 방법을 고객에게 안내해야 합니다.",
            "Final Answer: 고객님, 주문 관리 페이지에서 배송 알림을 설정하시면, 주문이 배송 완료되었을 때 알림을 받으실 수 있습니다. 설정 방법에 대한 자세한 안내를 원하시면 말씀해 주세요."
        ]


    for step in simulated_steps:
        print(step) # 시뮬레이션된 로그 출력
        logs.append(step)

    print("--- 시뮬레이션 완료 ---")
    return logs


def extract_tool_actions_from_logs(logs: List[str]) -> List[str]:
    """
    로그 라인에서 도구 액션의 순서를 추출합니다.
    "도구 이름[도구 입력값]" 형식의 문자열 목록을 반환합니다.
    """
    actions = []
    for line in logs:
        # ReAct AgentExecutor의 verbose 출력 형식에 맞게 파싱
        if line.strip().startswith("Action:"):
            # "Action: tool_name[tool_input]" 형식 파싱
            try:
                action_part = line.split('Action:')[1].strip()
                # 입력값이 없는 경우 (예: Action: Tool Name[]) 또는 있는 경우 처리
                if '[' in action_part and action_part.endswith(']'):
                     actions.append(action_part)
                else: # 예상치 못한 Action 형식
                     actions.append(f"Malformed Action: {action_part}")
            except IndexError:
                 actions.append(f"Malformed Action Line: {line}")
    return actions

def evaluate_performance_by_tool_sequence(logs: List[str], expected_sequence: List[str], scenario_type: str) -> Dict[str, Any]:
    """
    로그에서 추출된 도구 액션 순서를 기반으로 에이전트 성능을 평가하고,
    다양한 시나리오에 대한 예상 순서와 비교합니다.

    점수:
    - 100점: 예상된 정확한 액션 순서를 따랐습니다.
    - 80점: 핵심 도구를 논리적인 순서로 사용했지만, 사소한 편차나 추가적인 관련 단계가 있을 수 있습니다.
    - 50점: 일부 관련 도구를 사용했지만, 주요 단계가 누락되었거나 순서가 크게 틀렸습니다.
    - 20점: 관련 도구를 거의 사용하지 않았습니다.
    - 0점: 에이전트가 충돌했거나 의미 없는 출력을 생성했습니다 (이 간단한 평가기에서는 처리되지 않음).

    '배송 지연 및 쿠폰' 시나리오에 대한 특정 검사를 포함하여
    올바른 논리적 흐름(주문 상태 확인 -> 정책 검색 -> 쿠폰 발급)에 점수를 부여합니다.
    """
    actual_actions = extract_tool_actions_from_logs(logs)
    score = 0
    feedback = []

    print(f"\n--- 성능 평가 중 ({scenario_type}) ---")
    print(f"예상 도구 순서: {expected_sequence}")
    print(f"실제 도구 순서:   {actual_actions}")

    # 시나리오별 평가 로직
    if scenario_type == "Delayed Shipping & Coupon":
        # 예상 논리적 흐름: 상태 확인 -> 정책 검색 -> 쿠폰 발급 (지연된 경우)
        # 참고: 단순화된 검사이며, 실제로는 Observation 결과가 다음 단계에 영향을 미치는지 확인해야 합니다.
        if actual_actions == ["Get Order Status[ORDER456]", "Policy Search[배송 지연 보상 정책]", "Issue Complaint Coupon[ORDER456]"]:
            score = 100
            feedback.append("배송 지연 및 쿠폰에 대한 완벽한 순서입니다.")
        elif "Get Order Status[ORDER456]" in actual_actions and "Policy Search[배송 지연 보상 정책]" in actual_actions and "Issue Complaint Coupon[ORDER456]" in actual_actions:
             score = 80 # 모든 핵심 도구를 사용했지만, 순서가 다르거나 추가 단계가 있을 수 있습니다.
             feedback.append("주문 상태 확인, 정책 검색, 쿠폰 발급 도구를 사용했습니다.")
        elif "Get Order Status[ORDER456]" in actual_actions and "Policy Search[배송 지연 보상 정책]" in actual_actions:
             score = 70 # 상태 및 정책은 확인했지만, 쿠폰을 발급하지 않았습니다 (정책 확인 실패 또는 상태가 지연되지 않은 경우 올바를 수 있지만, 모의에서는 쿠폰 발급으로 이어져야 한다고 가정).
             feedback.append("주문 상태와 정책은 확인했지만, 쿠폰 발급을 시도하지 않았습니다.")
        elif "Get Order Status[ORDER456]" in actual_actions:
             score = 50 # 상태만 확인했습니다.
             feedback.append("주문 상태만 확인했습니다.")
        else:
            score = 20 # 관련 도구를 사용하지 않았습니다.
            feedback.append("이 시나리오에 대한 관련 도구를 사용하지 않았습니다.")


    elif scenario_type in ["Simple Status Check (Delivered)", "Simple Status Check (Processing)", "Unknown Order ID"]:
        # 예상 순서: 상태만 확인
        if actual_actions == expected_sequence:
            score = 100
            feedback.append("간단한 상태 확인에 대한 올바른 순서입니다.")
        elif "Get Order Status" in [action.split('[')[0] for action in actual_actions]:
            score = 80 # 올바른 도구를 사용했지만, 추가 단계가 있을 수 있습니다.
            feedback.append("주문 상태 확인 도구를 사용했지만, 순서에 편차가 있을 수 있습니다.")
        else:
            score = 20 # 상태 도구를 사용하지 않았습니다.
            feedback.append("주문 상태 확인 도구를 사용하지 않았습니다.")

    elif scenario_type == "Refund Policy Query": # 보상 정책 쿼리에서 변경됨
        # 예상 순서: 정책 검색 사용
        if actual_actions == expected_sequence:
             score = 100
             feedback.append("정책 검색에 대한 올바른 순서입니다.")
        elif "Policy Search" in [action.split('[')[0] for action in actual_actions]:
             score = 80 # 올바른 도구를 사용했지만, 추가 단계가 있을 수 있습니다.
             feedback.append("정책 검색 도구를 사용했지만, 순서에 편차가 있을 수 있습니다.")
        else:
             score = 20 # 정책 검색 도구를 사용하지 않았습니다.
             feedback.append("정책 검색 도구를 사용하지 않았습니다.")
    elif scenario_type == "Multi-turn Policy Follow-up": # 멀티턴 테스트를 위한 새로운 시나리오
        # 예상 순서: 정책 검색
        if actual_actions == expected_sequence:
             score = 100
             feedback.append("멀티턴 정책 후속 질문에 대한 올바른 순서입니다.")
        elif "Policy Search" in [action.split('[')[0] for action in actual_actions]:
             score = 80 # 멀티턴에서 올바른 도구를 사용했지만, 추가 단계가 있을 수 있습니다.
             feedback.append("멀티턴에서 정책 검색 도구를 사용했지만, 순서에 편차가 있을 수 있습니다.")
        else:
             score = 20 # 멀티턴에서 정책 검색 도구를 사용하지 않았습니다.
             feedback.append("멀티턴에서 정책 검색 도구를 사용하지 않았습니다.")


    else:
        # 예상치 못한 시나리오에 대한 기본 점수
        if actual_actions == expected_sequence:
            score = 100
            feedback.append("예상된 정확한 순서를 따랐습니다.")
        elif len(actual_actions) > 0:
            score = 50 # 일부 도구를 사용했습니다.
            feedback.append("일부 도구를 사용했지만, 순서나 관련성이 명확하지 않습니다.")
        else:
            score = 10 # 도구 사용 없음
            feedback.append("도구 사용이 감지되지 않았습니다.")


    # 예상치 못한 액션에 대한 일반 피드백
    # 참고: 이는 단순화된 검사입니다. 실제 시스템은 더 정교한 분석이 필요합니다.
    expected_tool_names_flat = [tool_with_input.split('[')[0] for tool_with_input in expected_sequence]
    actual_tool_names_flat = [action.split('[')[0] for action in actual_actions]

    for actual_tool_name in actual_tool_names_flat:
        # agent_tools 목록에 도구가 존재하는지 확인합니다 (이 범위에서 사용 가능하다고 가정).
        if 'agent_tools' in globals() and actual_tool_name not in expected_tool_names_flat and actual_tool_name in [t.name for t in agent_tools]:
             feedback.append(f"예상치 못한 도구 사용: {actual_tool_name}")
             if score > 20: score = max(20, score - 5) # 예상치 못한 도구 사용에 대한 사소한 패널티

    # 누락된 예상 액션 확인
    for expected_action in expected_sequence:
        if expected_action not in actual_actions:
             feedback.append(f"누락된 예상 액션: {expected_action}")
             if score > 20: score = max(20, score - 10) # 주요 단계 누락에 대한 패널티


    if len(actual_actions) > len(expected_sequence) and score > 50:
        feedback.append(f"에이전트가 {len(actual_actions) - len(expected_sequence)}개의 추가 액션을 수행했습니다.")


    print(f"점수: {score}")
    print(f"피드백: {feedback}")
    print("--------------------------")

    return {"score": score, "feedback": feedback, "actual_sequence": actual_actions, "scenario": scenario_type}

# 2. evaluate_performance 함수 개선
# 위 함수 정의 내에서 완료되었습니다.

# --- 사용 예시 ---
# agent_executor_with_memory가 사용 가능하거나 시뮬레이션 함수를 사용한다고 가정합니다.
# get_order_status, issue_complaint_coupon, agent_tools가 정의되었다고 가정합니다.

# 시뮬레이션을 실행하려면 실제 agent_executor가 필요하지 않지만,
# 평가 함수는 run_agent_and_capture_logs에서 캡처된 가정된 동작에 의존합니다.

# 시나리오 1: 성공적인 배송 지연 -> 정책 검색 -> 쿠폰 발급
query_delayed = "주문 ID ORDER456의 배송 상태가 어떻게 되나요? 만약 지연되었다면 쿠폰을 받을 수 있나요?"
# 이 시나리오에 대한 예상 순서: 상태 확인, 지연된 경우 쿠폰 발급
expected_seq_delayed = ["Get Order Status[ORDER456]", "Policy Search[배송 지연 보상 정책]", "Issue Complaint Coupon[ORDER456]"] # 예상 순서 업데이트

print("\n시나리오 1 평가 중: 배송 지연 및 쿠폰 요청")
# 이 시나리오에 대한 로그를 얻기 위해 시뮬레이션 함수를 사용합니다.
# 실제 시나리오에서는 agent_executor_with_memory.invoke(query_delayed, chat_history=[])를 사용합니다.
logs_delayed = run_agent_and_capture_logs(None, query_delayed, [])
evaluation_delayed = evaluate_performance_by_tool_sequence(logs_delayed, expected_seq_delayed, "Delayed Shipping & Coupon")
print(f"평가 결과 (시나리오 1): {evaluation_delayed}")


# 시나리오 2: 간단한 상태 확인 (배송 완료)
query_delivered = "주문 ID ORDER123의 배송 상태를 알려주세요."
# 예상 순서: 상태만 확인
expected_seq_delivered = ["Get Order Status[ORDER123]"]

print("\n시나리오 2 평가 중: 간단한 상태 확인 (배송 완료)")
# 이 시나리오에 대한 로그를 얻기 위해 시뮬레이션 함수를 사용합니다.
logs_delivered = run_agent_and_capture_logs(None, query_delivered, [])
evaluation_delivered = evaluate_performance_by_tool_sequence(logs_delivered, expected_seq_delivered, "Simple Status Check (Delivered)")
print(f"평가 결과 (시나리오 2): {evaluation_delivered}")

# 시나리오 3: 간단한 상태 확인 (처리 중)
query_processing = "주문 ID ORDER789 상태 확인 부탁드려요."
# 예상 순서: 상태만 확인
expected_seq_processing = ["Get Order Status[ORDER789]"]

print("\n시나리오 3 평가 중: 간단한 상태 확인 (처리 중)")
# 이 시나리오에 대한 로그를 얻기 위해 시뮬레이션 함수를 사용합니다.
logs_processing = run_agent_and_capture_logs(None, query_processing, [])
evaluation_processing = evaluate_performance_by_tool_sequence(logs_processing, expected_seq_processing, "Simple Status Check (Processing)")
print(f"평가 결과 (시나리오 3): {evaluation_processing}")


# 시나리오 4: 알 수 없는 주문 ID
query_unknown = "주문 ID ORDER_UNKNOWN 상태 알려줘."
# 예상 순서: 상태만 확인
expected_seq_unknown = ["Get Order Status[ORDER_UNKNOWN]"]

print("\n시나리오 4 평가 중: 알 수 없는 주문 ID")
# 이 시나리오에 대한 로그를 얻기 위해 시뮬레이션 함수를 사용합니다.
logs_unknown = run_agent_and_capture_logs(None, query_unknown, [])
evaluation_unknown = evaluate_performance_by_tool_sequence(logs_unknown, expected_seq_unknown, "Unknown Order ID")
print(f"평가 결과 (시나리오 4): {evaluation_unknown}")


# 시나리오 5: 정책 쿼리 (환불)
query_refund_policy = "환불 정책에 대해 알려줄 수 있나요?"
# 예상 순서: 정책 검색 사용
expected_seq_refund = ["Policy Search[환불 정책에 대해 알려줄 수 있나요?]"] # 변수 이름 변경

print("\n시나리오 5 평가 중: 환불 정책 쿼리 (시뮬레이션 RAG)")
# 이 시나리오에 대한 로그를 얻기 위해 시뮬레이션 함수를 사용합니다.
logs_refund = run_agent_and_capture_logs(None, query_refund_policy, [])
evaluation_refund = evaluate_performance_by_tool_sequence(logs_refund, expected_seq_refund, "Refund Policy Query")
print(f"평가 결과 (시나리오 5): {evaluation_refund}")

# 시나리오 6: 멀티턴 후속 질문 (시뮬레이션)
# chat history를 시뮬레이션해야 합니다. run_agent_and_capture_logs 함수는
# 쿼리와 chat_history 확인을 기반으로 이를 시뮬레이션하는 기본적인 기능을 가지고 있습니다.
# 예: 사용자가 상태를 묻고(턴 1), "그 주문"의 알림에 대해 묻습니다(턴 2).
# (개념적으로) chat history를 구축하기 위해 턴 1을 먼저 시뮬레이션합니다.
# 그런 다음 시뮬레이션된 chat history를 전달하여 후속 쿼리로 턴 2를 시뮬레이션합니다.
query_multi_turn_follow_up = "그 주문이 배송 완료되면 알려줄 수 있나요?"
# 이전 턴에서 ORDER123이 논의된 chat history를 시뮬레이션합니다.
simulated_chat_history = [
    HumanMessage(content="주문 ID ORDER123의 상태를 알려주세요."),
    AIMessage(content="주문 ID ORDER123의 배송 상태는 'Delivered' 입니다.") # 에이전트의 이전 응답 시뮬레이션
]
expected_seq_multi_turn = ["Policy Search[배송 완료 알림 정책]"] # 쿼리에 기반한 예상 액션

print("\n시나리오 6 평가 중: 멀티턴 정책 후속 질문 (시뮬레이션)")
# 이 시나리오에 대한 로그를 얻기 위해 시뮬레이션 함수를 사용하고, 시뮬레이션된 chat history를 전달합니다.
logs_multi_turn = run_agent_and_capture_logs(None, query_multi_turn_follow_up, simulated_chat_history)
evaluation_multi_turn = evaluate_performance_by_tool_sequence(logs_multi_turn, expected_seq_multi_turn, "Multi-turn Policy Follow-up")
print(f"평가 결과 (시나리오 6): {evaluation_multi_turn}")


# 랩업 보고서 생성 (개념)
print("\n## 에이전트 성능 랩업 보고서 (개념) ##")
print("이곳에 위 평가 결과를 바탕으로 에이전트의 전반적인 성능에 대한 보고서를 작성할 수 있습니다.")
print("포함될 수 있는 내용:")
print("- 각 시나리오별 에이전트의 도구 사용 순서 및 논리성 평가 점수")
print("- 에이전트가 ReAct 패턴을 잘 따랐는지 여부 (Thought, Action, Observation)")
print("- 멀티턴 대화에서 이전 맥락을 잘 활용했는지 여부 (메모리)")
print("- 신뢰성 로직 (예: 쿠폰 발급 조건) 준수 여부")
print("- JSON 출력 포맷 준수 여부 (과제 3 결과 바탕)")
print("- 개선이 필요한 부분 및 향후 계획")
print("\n실제 보고서는 평가 결과를 취합하고 분석하여 상세하게 작성해야 합니다.")
```

*   **Tool-learning (도구 학습):** Agent에게 외부 함수나 시스템과 상호작용할 수 있는 도구를 제공하고 사용하는 방법을 배웠습니다. (정책 검색, 쿠폰 발급 도구)
*   **RAG (검색 증강 생성):** 외부 문서(정책 파일)를 Agent의 지식 기반으로 확장하고, 이를 바탕으로 정확한 답변을 생성하는 방법을 이해했습니다. (정책 검색 도구 확장)
*   **Trustworthiness (신뢰성):** Agent가 민감한 도구(예: 쿠폰 발급)를 사용할 때 특정 조건을 확인하여 안전하고 신뢰할 수 있게 작동하도록 제어하는 로직을 구현했습니다.
*   **ReAct (Reasoning and Acting):** Agent가 Thought(추론), Action(행동), Observation(관찰 결과) 단계를 반복하며 복잡한 문제를 해결하는 논리적인 사고 과정을 시스템 프롬프트를 통해 유도하는 방법을 학습했습니다.
*   **Planning & Reasoning:** ReAct 프레임워크를 통해 Agent가 스스로 문제 해결을 위한 계획을 세우고 추론하는 능력을 향상시키는 개념을 체득했습니다.
*   **MCP (Model Context Protocol) & JSON 출력:** 다른 시스템과 표준화된 형식(JSON)으로 데이터를 교환하기 위한 데이터 모델(Pydantic)을 정의하고, Agent가 이러한 구조화된 출력을 생성하도록 유도하는 방법을 경험했습니다. 이는 멀티 에이전트 상호작용의 기초입니다.
*   **Memory (메모리):** Agent가 이전 대화 내용을 기억하여 멀티턴 대화의 맥락을 유지하고 자연스러운 상호작용을 할 수 있도록 메모리 기능을 Agent 실행 체인에 통합했습니다.
*   **Agent 성능 평가:** Agent의 문제 해결 과정, 특히 도구 사용의 논리적 순서를 기반으로 Agent의 성능을 평가하는 접근 방식(Reward Model 개념 연결)을 살펴보았습니다.

이 실습을 통해 여러분은 기본적인 AI 에이전트가 실제 서비스 환경에서 복합적인 고객 민원을 처리하고, 외부 시스템과 협력하며, 대화 맥락을 유지하는 '똑똑한' 에이전트로 발전하는 과정에 대해 이해하셨기를 바랍니다.
